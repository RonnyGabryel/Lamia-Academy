# Linear Regression

Let's fabricate some data that shows a roughly linear relationship between page speed and amount purchased:

```python
import numpy as np
from matplotlib.pyplot import scatter
import matplotlib.pyplot as plt
```

```python
pageSpeeds = np.random.normal(3.0, 1.0, 1000)  # ele gera 1000 valores aleatórios, para representar a velocidade da pagina
purchaseAmount = 100 - (pageSpeeds + np.random.normal(0, 0.1, 1000)) * 3  # faz o calculo do valor de compras com base na velocidade da página

scatter(pageSpeeds, purchaseAmount)  # Cria um gráfico de dispersão (scatter plot) para visualizar a relação entre velocidades de página e valores de compra
```

As we only have two features, we can keep it simple and just use scipy.state.linregress:

```python
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(pageSpeeds, purchaseAmount) # Regressão linear
```

Not surprisngly, our R-squared value shows a really good fit:

```python
r_value ** 2 # Coeficiente de determinação
```

Let's use the slope and intercept we got from the regression to plot predicted values vs. observed:

```python
import matplotlib.pyplot as plt

# Função para prever valores
def predict(x):
    return slope * x + intercept # y = mx + b

fitLine = predict(pageSpeeds)

# Gráfico de dispersão com a linha de regressão
plt.scatter(pageSpeeds, purchaseAmount)
plt.plot(pageSpeeds, fitLine, c='r')
plt.show()
```
