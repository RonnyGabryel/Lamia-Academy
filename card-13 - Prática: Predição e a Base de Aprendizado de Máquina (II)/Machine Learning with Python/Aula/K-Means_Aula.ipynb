# K-Means Clustering Example
Let's make some fake data that includes people clustered by income and age, randomly:

```python
from numpy import random, array

def createClusteredData(N, k):#criar grupos de n pessoas
    np.random.seed(10)#uma semente aletorio 
    pointsPerCluster = float(N)/k # Define os números de pontos por cluster
    X = []
    for i in range(k):#Cria os clusters
        incomeCentroid = np.random.uniform(20000.0, 200000.0) # Define o centro de renda entre 20k e 200k
        ageCentroid = np.random.uniform(20.0, 70.0) # Define o centro de idade entre 20 e 70 anos
        for j in range(int(pointsPerCluster)): # Cria os pontos
            X.append([np.random.normal(incomeCentroid, 10000.0), np.random.normal(ageCentroid, 2.0)]) # Cria os pontos com base no centro e desvio padrão
    X = array(X)
    return X
```

We'll use k-means to rediscover these clusters in unsupervised learning:

```python
%matplotlib inline

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from numpy import random

data = createClusteredData(100, 5) #100 pessoas em 5 grupos direntes

model = KMeans(n_clusters=5)

# Note I'm scaling the data to normalize it! Important for good results.
model = model.fit(scale(data))

# We can look at the clusters each data point was assigned to
print(model.labels_)

# And we'll visualize it:
plt.figure(figsize=(8, 6)) #criando o grid
plt.scatter(x=data[:,0], y=data[:,1], c=model.labels_.astype(float)) #atribui cor aos rotuloscriados
plt.show()
```
