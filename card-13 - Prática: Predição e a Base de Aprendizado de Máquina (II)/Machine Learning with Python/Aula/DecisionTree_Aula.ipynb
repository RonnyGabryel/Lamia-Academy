# Decision TreesFirst 
First we'll load some fake data on past hires I made up. Note how we use pandas to convert a csv file into a DataFrame:

```python
import numpy as np
import pandas as pd
from sklearn import tree

input_file = "/home/Ront/Documentos/Card-13/PastHires.csv"
df = pd.read_csv(input_file, header = 0)
```

```python
df.head()
```

scikit-learn needs everything to be numerical for decision trees to work. So, we'll map Y/N to 1,0 and levels of education to some scale of 0-2. In the real world, you'd need to think about how to deal with unexpected or missing data! By using map(), we know we'll get NaN for unexpected values.

```python
# Invertendo a árvore de decisão
d = {'Y': 0, 'N': 1}
df['Hired'] = df['Hired'].map(d)
d = {'Y': 1, 'N': 0}
df['Employed?'] = df['Employed?'].map(d)
df['Top-tier school'] = df['Top-tier school'].map(d)
df['Interned'] = df['Interned'].map(d)
d = {'BS': 0, 'MS': 1, 'PhD': 2}
df['Level of Education'] = df['Level of Education'].map(d)
df.head()
```

Next we need to separate the features from the target column that we're trying to build a decsion tree for.

```python
features = list(df.columns[:6]) # Pegando as 6 primeiras colunas
features
```

Now actualy construct the decison tree:

```python
y = df["Hired"] #Coluna Hired
X = df[features]#Colunas de features
clf = tree.DecisionTreeClassifier()# nstanciando o classificador
clf = clf.fit(X,y) # Treinando o classificador
```


... and display it. Note you need to have pyplot2 instaled for this to work.
To read this decison tree, each condition branches left for "true" and right for "false". When you end up at a value, the value array represents how many samples exist in each target value. So value = [0. 5.] mean there are 0 "no hires" and 5 "hires" by the tim we get to that point. value = [3. 0.] means 3 no-hires and 0 hires.

```python
from IPython.display import Image
from six import StringIO
import pydot

dot_data = StringIO()
tree.export_graphviz(clf, out_file=dot_data,
                    feature_names=features)
graph = pydot.graph_from_dot_data(dot_data.getvalue())[0]
Image(graph.create_png())
```

# Ensemble learning: using a random forest
We'll use a random forest of 10 decison trees to predict employment of specific candidate profiles:

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=10)
clf = clf.fit(X, y)

#Predict employment of an employed 10-year veteran
print(clf.predict([[10, 1, 4, 0, 0, 0]]))
#Predict employment of an unemployed 10-year veteran
print(clf.predict([[10, 0, 4, 0, 0, 0]]))
```
