# Activity
Try different polynomial orders. Can you get a better fit with higher orders? Do you start to see overfitting, even though the r-squared score looks good for this particular data set?

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

```python
# Gera dados
np.random.seed(2)
pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = np.random.normal(50.0, 10.0, 1000) / pageSpeeds

# Converte para arrays só pra deixar explícito
x = np.array(pageSpeeds)
y = np.array(purchaseAmount)
```

```python
# Cria uma grade 2x3 de gráficos
# Um gráfico para cada grau do polinômio
fig, axes = plt.subplots(2, 3, figsize=(14, 8))
axes = axes.ravel()

# Pontos usados só para desenhar a curva suave
xp = np.linspace(0, 7, 100)

for i, grau in enumerate(range(1, 7)):
    p = np.poly1d(np.polyfit(x, y, grau))
    # Calcula o R usando os mesmos dados
    r2 = r2_score(y, p(x))
    
    # Plota os pontos reais
    axes[i].scatter(x, y, alpha=0.3, s=8)
    axes[i].plot(xp, p(xp), 'r-', linewidth=2)
    axes[i].set_title(f'Grau {grau}: R² = {r2:.4f}')
    axes[i].grid(True, alpha=0.3)
    
    print(f"Grau {grau}: R² = {r2:.4f}")

plt.tight_layout()
plt.show()
```
