```python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer, Dense, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras import utils as np_utils
import numpy as np
from sklearn.model_selection import StratifiedKFold #
```

```python
(X, y), (X_teste, y_teste) = mnist.load_data()
X = X.reshape(X.shape[0], 28, 28, 1) #quantidade de resgistros
X = X.astype('float32') #conversГЈo para tipo float32
X /= 255 #a divisГЈo por 255 para os pixel ficar na faixa 0 e 1
y = np_utils.to_categorical(y, 10) #conversГЈo dos labels de treino para one hot
```

```python
np.zeros(shape = (y.shape[0], 1)) #matriz de zeros com o mesmo nГєmero de linhas
```

```python
seed = 5
np.random.seed(seed) # Usando a seed para garantir a reprodutibilidade dos resultados
```

```python
kfold = StratifiedKFold(n_splits = 5, shuffle=True, random_state = seed) #fazendo assim para ele memorizar
```

```python
# praticamente a mesma coisa do MNIST Simples
resultados = []
for indice_treinamento, indice_teste in kfold.split(X, np.zeros(shape = (y.shape[0], 1))):
  print("ГЌndices treinamento: ", indice_treinamento, "ГЌndice teste: ", indice_teste)
  classificador = Sequential()
  #Camada de convoluГ§ГЈo
  classificador.add(InputLayer(shape=(28, 28, 1)))
  classificador.add(Conv2D(32, (3, 3), activation='relu'))
  classificador.add(MaxPooling2D(pool_size=(2,2)))
  classificador.add(Flatten())
  # Camada oculta
  classificador.add(Dense(units=128, activation='relu'))
  # Camada saГ­da
  classificador.add(Dense(units=10, activation='softmax'))
  classificador.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
  classificador.fit(X[indice_treinamento], y[indice_treinamento], batch_size = 128, epochs = 5) #usadndo dados de treino com o Г­ndice de treinamento
  precisao = classificador.evaluate(X[indice_teste], y[indice_teste]) # usando dados de teste com o Г­ndice de teste
  resultados.append(precisao[1])
```
