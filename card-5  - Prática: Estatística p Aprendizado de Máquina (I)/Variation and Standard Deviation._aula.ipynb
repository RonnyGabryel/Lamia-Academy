```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Experimento 1: Desvio padrão pequeno
print("Experimento 1: Desvio padrão = 10")
incomes1 = np.random.normal(100.0, 10.0, 10000)
plt.hist(incomes1, 50, color='green')
plt.title('Distribuição com Desvio Padrão = 10')
plt.show()
print(f"Desvio padrão: {incomes1.std():.2f}")
print(f"Variância: {incomes1.var():.2f}")
```

```python
# Experimento 2: Desvio padrão médio (do código original)
print("Experimento 2: Desvio padrão = 20")
incomes2 = np.random.normal(100.0, 20.0, 10000)
plt.hist(incomes2, 50, color='red')
plt.title('Distribuição com Desvio Padrão = 20')
plt.show()
print(f"Desvio padrão: {incomes2.std():.2f}")
print(f"Variância: {incomes2.var():.2f}")
print()
```

```python
# Experimento 3: Desvio padrão grande (do código da imagem)
print("Experimento 3: Desvio padrão = 50")
incomes3 = np.random.normal(100.0, 50.0, 10000)
plt.hist(incomes3, 50, color='orange')
plt.title('Distribuição com Desvio Padrão = 50')
plt.show()
print(f"Desvio padrão: {incomes3.std():.2f}")
print(f"Variância: {incomes3.var():.2f}")
```
