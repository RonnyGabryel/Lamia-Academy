```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def de_mean(x):
    xmean = np.mean(x)
    return [xi - xmean for xi in x]
```

```python
def covariance(x, y):
    n = len(x)
    return np.dot(de_mean(x), de_mean(y)) / (n-1)
```

```python
def correlation(x, y):
    stddevx = x.std()
    stddevy = y.std()
    return covariance(x, y) / stddevx / stddevy
```

```python
# Dados base
pageSpeeds = np.random.normal(3.0, 1.0, 1000)
```

```python
# Experimento 1: Dados independentes
purchaseAmount1 = np.random.normal(50.0, 10.0, 1000)
plt.scatter(pageSpeeds, purchaseAmount1)
plt.title("Experimento 1: Independentes")
plt.show()
print("Covarincia:", covariance(pageSpeeds, purchaseAmount1))
print("Correlao:", correlation(pageSpeeds, purchaseAmount1))
print("NumPy:", np.corrcoef(pageSpeeds, purchaseAmount1)[0,1])
print()
```

```python
# Experimento 2: Relao inversa com rudo
purchaseAmount2 = np.random.normal(50.0, 10.0, 1000) / pageSpeeds
plt.scatter(pageSpeeds, purchaseAmount2)
plt.title("Experimento 2: Inversa com rudo")
plt.show()
print("Covarincia:", covariance(pageSpeeds, purchaseAmount2))
print("Correlao:", correlation(pageSpeeds, purchaseAmount2))
print("NumPy:", np.corrcoef(pageSpeeds, purchaseAmount2)[0,1])
print()
```

```python
# Experimento 3: Relao linear perfeita
purchaseAmount3 = 100 - pageSpeeds * 3
plt.scatter(pageSpeeds, purchaseAmount3)
plt.title("Experimento 3: Linear perfeita")
plt.show()
print("Covarincia:", covariance(pageSpeeds, purchaseAmount3))
print("Correlao:", correlation(pageSpeeds, purchaseAmount3))
print("NumPy:", np.corrcoef(pageSpeeds, purchaseAmount3)[0,1])
```
