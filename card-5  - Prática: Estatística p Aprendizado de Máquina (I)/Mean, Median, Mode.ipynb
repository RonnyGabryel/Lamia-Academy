```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# Gerando 6000 notas de estudantes (0 a 100)
# Distribuição normal com média 75 e desvio padrão 12
notas = np.random.normal(75, 12, 6000)
```

```python
# Limitando as notas entre 0 e 100 (mais realista)
notas = np.clip(notas, 0, 100)
```

```python
# Calculando estatísticas iniciais
print("Média inicial das notas:", np.mean(notas))
```

```python
# Visualizando as notas originais
plt.hist(notas, 50, color='green') # Coloquei a cor de verder por perferencia minha
plt.title('Distribuição das Notas')
plt.show()
```

```python
# Calculando mediana inicial
print("Mediana inicial das notas:", np.median(notas))
```

```python
# Adicionando algumas notas muito baixas (outliers)
# Simulando estudantes que faltaram às provas
notas_com_outliers = np.append(notas, [5, 8, 12, 3, 15])
```

```python
# Novas estatísticas após outliers
print("Mediana após outliers:", np.median(notas_com_outliers))
print("Média após outliers:", np.mean(notas_com_outliers))
```

```python
# Exemplo com idades de funcionários
idades_funcionarios = np.random.randint(22, high=65, size=800)
print("Idades dos funcionários:", idades_funcionarios)
```

```python
# Calculando a moda das idades
from scipy import stats
resultado_moda = stats.mode(idades_funcionarios)
print("Moda das idades:", resultado_moda)
```

```python
# Gerando dados de salários
# Distribuição normal com média 5000 e desvio padrão 1500
salarios = np.random.normal(5000, 1500, 4000)
```

```python
# Visualizando salários
plt.hist(salarios, 40, color='green')
plt.title('Distribuição dos Salários')
plt.show()
```

```python
# Estatísticas dos salários
print("Mediana dos salários:", np.median(salarios))
print("Média dos salários:", np.mean(salarios))
```

```python
# Adicionando salário de um CEO (outlier extremo)
salarios_com_ceo = np.append(salarios, [500000])
```

```python
# Comparando após adicionar o CEO
print("Mediana após CEO:", np.median(salarios_com_ceo))
print("Média após CEO:", np.mean(salarios_com_ceo))
```
