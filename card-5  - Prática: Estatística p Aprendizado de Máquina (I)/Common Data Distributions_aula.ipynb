```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, expon, binom, poisson
```

```python
# Distribuição Uniforme
print("\n1. Distribuição Uniforme")
print("Todos os valores têm a mesma chance de ocorrer")

values = np.random.uniform(-10.0, 10.0, 100000)
plt.figure(figsize=(8, 5))
plt.hist(values, 50, alpha=0.7, color='green')
plt.title('Distribuição Uniforme')
plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Distribuição Normal - PDF teórica
print("\n2. Distribuição Normal (Curva Teórica)")
print("A famosa curva em formato de sino")

x = np.arange(-4, 4, 0.01)
y = norm.pdf(x, 0, 1)
plt.figure(figsize=(8, 5))
plt.plot(x, y, linewidth=2)
plt.fill_between(x, y, alpha=0.3)
plt.title('Distribuição Normal Padrão (μ=0, σ=1)')
plt.xlabel('Valores')
plt.ylabel('Densidade de Probabilidade')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Distribuição Normal - Amostragem
print("\n3. Distribuição Normal (Amostra)")
print("Gerando dados que seguem a distribuição normal")

mu, sigma = 5.0, 2.0
values = np.random.normal(mu, sigma, 10000)
plt.figure(figsize=(8, 5))
plt.hist(values, 50, alpha=0.7, color='green')
plt.title(f'Amostra Normal (μ={mu}, σ={sigma})')
plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Distribuição Exponencial
print("\n4. Distribuição Exponencial")
print("Útil para modelar tempo entre eventos")

x = np.arange(0, 6, 0.01)
y = expon.pdf(x, scale=1)
plt.figure(figsize=(8, 5))
plt.plot(x, y, linewidth=2)
plt.fill_between(x, y, alpha=0.3)
plt.title('Distribuição Exponencial')
plt.xlabel('Valores')
plt.ylabel('Densidade de Probabilidade')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Distribuição Binomial
print("\n5. Distribuição Binomial")
print("Número de sucessos em tentativas independentes")

n, p = 20, 0.3
x = np.arange(0, n+1)
y = binom.pmf(x, n, p)
plt.figure(figsize=(8, 5))
plt.bar(x, y, alpha=0.7)
plt.title(f'Distribuição Binomial (n={n}, p={p})')
plt.xlabel('Número de Sucessos')
plt.ylabel('Probabilidade')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Distribuição de Poisson
print("\n6. Distribuição de Poisson")
print("Número de eventos em intervalo fixo de tempo")

lam = 8
x = np.arange(0, 20)
y = poisson.pmf(x, lam)
plt.figure(figsize=(8, 5))
plt.bar(x, y, alpha=0.7)
plt.title(f'Distribuição de Poisson (λ={lam})')
plt.xlabel('Número de Eventos')
plt.ylabel('Probabilidade')
plt.grid(True, alpha=0.3)
plt.show()
```
