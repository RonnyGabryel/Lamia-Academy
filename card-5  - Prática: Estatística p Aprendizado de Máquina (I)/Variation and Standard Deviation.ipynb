```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Experimento 1: Desvio padr√£o pequeno
print("Experimento 1: Desvio padr√£o = 10")
incomes1 = np.random.normal(100.0, 10.0, 10000)
plt.hist(incomes1, 50, color='green')
plt.title('Distribui√ß√£o com Desvio Padr√£o = 10')
plt.show()
print(f"Desvio padr√£o: {incomes1.std():.2f}")
print(f"Vari√¢ncia: {incomes1.var():.2f}")
```

```python
# Experimento 2: Desvio padr√£o m√©dio (do c√≥digo original)
print("Experimento 2: Desvio padr√£o = 20")
incomes2 = np.random.normal(100.0, 20.0, 10000)
plt.hist(incomes2, 50, color='red')
plt.title('Distribui√ß√£o com Desvio Padr√£o = 20')
plt.show()
print(f"Desvio padr√£o: {incomes2.std():.2f}")
print(f"Vari√¢ncia: {incomes2.var():.2f}")
print()
```
