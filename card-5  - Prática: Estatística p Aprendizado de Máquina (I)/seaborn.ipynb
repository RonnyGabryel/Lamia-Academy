```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carrega os dados
df=pd.read_csv("http://media.sundog-soft.com/SelfDriving/FuelEfficiency.csv")
```

```python
# Scatter plot b√°sico
plt.scatter(df['# Gears'], df['CombMPG'])
plt.xlabel('Number of Gears')
plt.ylabel('Combined MPG')
plt.show()
```

```python
# Com seaborn fica mais bonito
sns.set()
plt.scatter(df['# Gears'], df['CombMPG'])
plt.xlabel('Number of Gears')
plt.ylabel('Combined MPG')
plt.show()
```

```python
# V√™ os dados
df[['# Gears', 'CombMPG']].head()
```

```python
# Scatter plot do seaborn
sns.scatterplot(x="# Gears", y="CombMPG", data=df)
plt.show()
```

```python
# Joint plot
sns.jointplot(x="# Gears", y="CombMPG", data=df)
plt.show()
```

```python
# Com linha de regress√£o
sns.lmplot(x="# Gears", y="CombMPG", data=df)
plt.show()
```

```python
# Box plot
sns.boxplot(x='# Gears', y='CombMPG', data=df)
plt.show()
```

```python
# Swarm plot
sns.swarmplot(x='# Gears', y='CombMPG', data=df)
plt.show()
```

```python
# V√™ as estat√≠sticas b√°sicas
print(df.groupby('# Gears')['CombMPG'].mean())
```

```python
# Heatmap
df2 = df.pivot_table(index='Cylinders', columns='Eng Displ', values='CombMPG', aggfunc='mean') # Novo DataFrame reorganizando o original
sns.heatmap(df2)
plt.show()
```
