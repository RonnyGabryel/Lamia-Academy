```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Gráfico básico de dispersão
age = np.random.randint(18, 65, 100)
tv_time = np.random.randint(1, 8, 100)
plt.scatter(age, tv_time)
plt.show()
```

```python
# Adicionando labels aos eixos
plt.scatter(age, tv_time)
plt.xlabel('Age (years)')
plt.ylabel('TV Time (hours/day)')
plt.show()
```

```python
# Versão com formatação melhorada
plt.figure(figsize=(8, 6))
plt.scatter(age, tv_time, alpha=0.6, s=50)
plt.xlabel('Age (years)')
plt.ylabel('TV Time (hours/day)')
plt.title('Age vs Time Spent Watching TV')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Versão com cores e tamanhos variados
colors = np.random.rand(100)
sizes = np.random.randint(20, 100, 100)
plt.figure(figsize=(10, 6))
plt.scatter(age, tv_time, c=colors, s=sizes, alpha=0.6, cmap='viridis')
plt.xlabel('Age (years)')
plt.ylabel('TV Time (hours/day)')  
plt.title('Age vs TV Time - Colorful Version')
plt.colorbar(label='Random Color Scale')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Dados mais realistas com tendência
np.random.seed(42)
age_realistic = np.random.normal(40, 15, 200)
age_realistic = np.clip(age_realistic, 18, 75)  # Limita entre 18-75 anos
```

```python
# Pessoas mais velhas tendem a assistir mais TV
tv_time_realistic = 2 + (age_realistic - 18) * 0.05 + np.random.normal(0, 1.5, 200)
tv_time_realistic = np.clip(tv_time_realistic, 0, 10)  # Limita entre 0-10 horas

plt.figure(figsize=(10, 6))
plt.scatter(age_realistic, tv_time_realistic, alpha=0.6, color='steelblue', s=40)
plt.xlabel('Age (years)')
plt.ylabel('TV Time (hours/day)')
plt.title('Age vs TV Time - Realistic Data with Trend')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
