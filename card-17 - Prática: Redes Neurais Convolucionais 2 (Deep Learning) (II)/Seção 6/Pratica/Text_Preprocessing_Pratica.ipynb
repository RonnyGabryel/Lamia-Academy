```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
# definindo dados de exemplo
sentencesias = [
    "To upando um monge WW pra brincar no PvP",
    "DK profano parece muito quebrado na arena",
    "Preciso aprender a dar kite melhor contra melee",
    "Battleground ranqueada deve ser uma loucura no WoW"
]
```

```python
# configurando
MAX_VOCAB = 20000 # tamanho maximo
MAX_LEN = 6
```

```python
tokenizer = Tokenizer(num_words=20000, oov_token="<OOV>") # se a palavra que não esteja no vocabulário aprendido sera substituída
tokenizer.fit_on_texts(sentencesias)
sequencias = tokenizer.texts_to_sequences(sentencesias)
data = pad_sequences(sequencias, maxlen=6)
```

```python
# transformando em sequências numéricas
sequencias = tokenizer.texts_to_sequences(sentencesias)
print(sequencias)
```

```python
tokenizer.word_index # Mostra o índice de cada palavra
```

```python
data = pad_sequences(sequencias) # realiza o padding das sequências
print(data)
```

```python
MAX_SEQUENCE_LENGTH = 5 # limitando para 5 palavras
data = pad_sequences(sequencias, maxlen=MAX_SEQUENCE_LENGTH)
print(data)
```

```python
data = pad_sequences(sequencias, maxlen=MAX_SEQUENCE_LENGTH, padding='post') # padding no final
print(data)
```

```python
data = pad_sequences(sequencias, maxlen=6) # padding com 6 de tamanho
print(data)
```

```python
data = pad_sequences(sequencias, maxlen=4) # aqui diminuímos o tamanho da sequência
print(data)
```

```python
data = pad_sequences(sequencias, maxlen=4, truncating='post') # truncando no final
print(data)
```
