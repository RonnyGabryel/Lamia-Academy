```python
import spacy # importa a biblioteca spacy
!python -m spacy download en_core_web_md
```

```python
nlp = spacy.load("en_core_web_md") # carrega o modelo medio
text = ("West Chestertenfieldville was referenced in Mr. Deeds.") # define o texto de teste
```

```python
doc = nlp(text) # processa o texto inicial
```

```python
for ent in doc.ents: # itera pelas entidades
    print(ent.text, ent.label) # exibe o texto e o id do rótulo
```

```python
ruler = nlp.get_pipe("entity_ruler") # acessa a ferramenta de regras
```

```python
nlp.analyze_pipes() # checa a estrutura
```

```python
patterns = [ # define os padrões que o spacy aprenda
    {"label": "GPE", "pattern": "West Chestertenfieldville"}]
```

```python
ruler.add_patterns(patterns) # injeta os novos padrões no ruler
```

```python
doc2 = nlp(text) # processa o mesmo texto com as novas regras
for ent in doc2.ents: # percorre as entidades do novo processamento
    print(ent.text, ent.label_) # exibe o rotulo atualizado
```

```python
nlp2 = spacy.load("en_core_web_sm") # carrega o modelo básico
```

```python
ruler = nlp2.add_pipe("entity_ruler", before = "ner") # coloca as regras
```

```python
ruler.add_patterns(patterns)
```

```python
doc = nlp2(text) # processa o texto
```

```python
for ent in doc.ents: # percorre as entidades achadas
    print(ent.text, ent.label) # exibe texto e id do rótulo
```

```python
nlp2.analyze_pipes()
```

```python
nlp3 = spacy.load("en_core_web_sm") # carrega um terceiro modelo
```

```python
ruler = nlp3.add_pipe("entity_ruler", before = "ner") # posiciona o ruler
```

```python
patterns = [ # define novos padroes
    {"label": "GPE", "pattern": "West Chestertenfieldville"},
    {"label": "FILM", "pattern": "Mr. Deeds"}
]
```

```python
ruler.add_patterns(patterns) # adiciona os padroes
```

```python
doc = nlp3(text) # processa o texto
```

```python
for ent in doc.ents: # itera sobre as entidades
    print(ent.text, ent.label) # mostra os resultados
```
