```python
df = pd.read_csv("stocks.tsv", delimiter="\t") # carrega o arquivo tsv
df # exibe o conteúdo da tabela carregada
```

```python
symbols = df.Symbol.tolist() # converte a coluna symbol do dataframe em uma lista
companies = df.CompanyName.tolist() # converte a coluna companyname em uma lista
print(symbols[:10]) # exibe os primeiros dez símbolos
```
