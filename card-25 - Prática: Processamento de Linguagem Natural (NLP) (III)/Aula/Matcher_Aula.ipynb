```python
import spacy # importa a biblioteca spacy
from spacy.matcher import Matcher # importa o matcher para busca
```

```python
nlp = spacy.load("en_core_web_sm") # carrega o modelo básico em inglês
```

```python
matcher = Matcher(nlp.vocab) # cria o buscador
pattern = [{"LIKE_EMAIL": True}] # define o padrão para identificar
matcher.add("EMAIL_ADDRESS", [pattern]) # adiciona o padrão
```

```python
doc = nlp("This is an email address: higormiler55@gmail.com") # processa o texto
matches = matcher(doc)
```

```python
print(matches) # exibe o id do padrao
```

```python
print(nlp.vocab[matches[0][0]].text) # converte o id numerico
```

```python
with open("wiki_mlk.txt") as f: # abre o arquivo sobre martin luther king
    text = f.read() # lê o conteúdo completo
```

```python
print(text) # imprime o texto para conferencia
```

```python
nlp = spacy.load("en_core_web_sm") # carrega o modelo básico
```

```python
matcher = Matcher(nlp.vocab) # cria o buscador
pattern = [{"POS": "PROPN"}] # define um padrao para capturar
matcher.add("PROPER_NOUN", [pattern]) # adiciona o padrão ao buscador
doc = nlp(text) # processa o texto carregado
matches = matcher(doc) # busca as ocorrências do padrão no texto
print(len(matches)) # exibe o total de combinações
for match in matches[:10]: # percorre apenas os 10 primeiros
    print(match, doc[match[1]:match[2]]) # exibe o id e o texto encontrado
```

```python
import json # importa a biblioteca para ler arquivos json
with open("alice.json", "r") as f: # abre o arquivo da alice no modo leitura
    data = json.load(f) # carrega os dados do json para a variável
```

```python
text = data[0][2][0] # seleciona uma parte específica da lista do json
print(text) # exibe o texto bruto
```

```python
text = text.replace("`","'") # troca a crase por aspas
print(text) # exibe o texto
```

```python
matcher = Matcher(nlp.vocab) # inicia o buscador do spacy
pattern = [{"ORTH": '"'}, # procura aspas iniciais
           {"IS_ALPHA": True, "OP": "+"}, # busca uma ou mais letras
           {"IS_PUNCT": True, "OP": "*"}, # busca pontuacoes se houver
           {"ORTH": '"'} # procura aspas finais
          ]

matcher.add("PROPER_NOUNS", [pattern], greedy='LONGEST') # define regra de busca longa
doc = nlp(text) # aplica o modelo ao texto
matches = matcher(doc) # encontra as coincidencias
matches.sort(key = lambda x: x[1]) # organiza pela ordem de texto
print (len(matches)) # imprime a quantidade de itens
for match in matches[:10]: # percorre os 10 primeiros
    print (match, doc[match[1]:match[2]])
```

```python
speak_lemmas = ["say", "think"] # lista de verbos para busca
matcher = Matcher(nlp.vocab) # inicia o buscador do spacy
pattern = [{"ORTH": "'"}, # procura aspas iniciais
           {"IS_ALPHA": True, "OP": "+"}, # busca uma ou mais letras
           {"IS_PUNCT": True, "OP": "*"}, # busca pontuacoes se houver
           {"ORTH": "'"}, # procura aspas finais
           {"POS": "VERB", "LEMMA": {"IN": speak_lemmas}}, # busca verbo da lista
           {"POS": "PROPN", "OP": "+"}, # busca um ou mais nomes proprios
           {"ORTH": "'"}, # busca outra aspa simples
           {"IS_ALPHA": True, "OP": "+"}, # busca mais letras
           {"IS_PUNCT": True, "OP": "*"}, # busca mais pontuacoes
           {"ORTH": "'"}, # busca aspa simples final
          ]

matcher.add("PROPER_NOUN_PATTERN", [pattern], greedy="LONGEST") # adiciona regra de busca longa
doc = nlp(text) # aplica o modelo ao texto
matches = matcher(doc) # encontra as coincidencias
matches.sort(key=lambda x: x[1]) # organiza pela ordem de texto
print(matches.__len__()) # imprime a quantidade de itens
for match in matches[:10]: # percorre os 10 primeiros
    print(match, doc[match[1]:match[2]])
```

```python
matcher = Matcher(nlp.vocab) # cria o objeto matcher
pattern1 = [{'ORTH': "'"}, {'IS_ALPHA': True, "OP": "+"}, {'IS_PUNCT': True, "OP": "*"}, {'ORTH': "'"}, {"POS": "VERB", "LEMMA": {"IN": speak_lemmas}}, {"POS": "PROPN", "OP": "+"}, {'ORTH': "'"}, {'IS_ALPHA': True, "OP": "+"}, {'IS_PUNCT': True, "OP": "*"}, {'ORTH': "'"}]
pattern2 = [{'ORTH': "'"}, {'IS_ALPHA': True, "OP": "+"}, {'IS_PUNCT': True, "OP": "*"}, {'ORTH': "'"}, {"POS": "VERB", "LEMMA": {"IN": speak_lemmas}}, {"POS": "PROPN", "OP": "+"}]
pattern3 = [{"POS": "PROPN", "OP": "+"},{"POS": "VERB", "LEMMA": {"IN": speak_lemmas}}, {'ORTH': "'"}, {'IS_ALPHA': True, "OP": "+"}, {'IS_PUNCT': True, "OP": "*"}, {'ORTH': "'"}]
matcher.add("PROPER_NOUNS", [pattern1, pattern2, pattern3], greedy='LONGEST') # adiciona os padrões ao matcher buscando a maior correspondência possível
for text in data[0][2]: # percorre os textos na lista de dados
    text = text.replace("`", "'") # padroniza o tipo de aspas no texto
    doc = nlp(text) # processa o texto com o modelo
    matches = matcher(doc) # encontra as correspondências no documento
    matches.sort(key = lambda x: x[1]) # ordena os resultados pela posição
    print (len(matches)) # exibe a quantidade
    for match in matches[:10]: # percorre os 10 primeiros resultados
        print (match, doc[match[1]:match[2]])
```
