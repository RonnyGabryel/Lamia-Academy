```python
nlp = spacy.load("en_core_web_md") # carrega o modelo médio para similaridade
```

```python
with open("wiki_us.txt") as f: # abre o arquivo de texto
    text = f.read() # lê o conteúdo do arquivo
```

```python
doc = nlp(text) # processa o texto no motor do spacy
sentence1 = list(doc.sents)[0] # pega a primeira frase da lista
print(sentence1) # exibe a frase na tela
```

```python
import numpy as np # importa o numpy para cálculos
your_word = "country" # define a palavra de entrada

# busca as 10 palavras mais parecidas no vocabulário
ms = nlp.vocab.vectors.most_similar(
    np.asarray([nlp.vocab.vectors[nlp.vocab.strings[your_word]]]), n=10)
words = [nlp.vocab.strings[w] for w in ms[0][0]] # transforma os ids em palavras
distances = ms[2] # guarda as distâncias matemáticas
print(words)
```

```python
doc1 = nlp("I like salty fries and hamburgues.") # processa frase 1
doc2 = nlp("Fast food eastes very good.") # processa frase 2
```

```python
print(doc1, "<->", doc2, doc1.similarity(doc2)) # compara similaridade entre elas
```

```python
doc3 = nlp("The Cristiano Ronaldo is the best player.") # processa frase 3
```

```python
print(doc1, "<->", doc3, doc1.similarity(doc3)) # compara temas diferentes
```
