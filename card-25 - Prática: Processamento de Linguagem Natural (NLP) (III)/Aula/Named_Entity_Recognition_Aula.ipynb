```python
text = "Higor enjoys playing football." # define o texto da variavel
doc2 = nlp(text) # processa o texto usando o modelo
print(doc2)
```

```python
for token in doc2: # percorre cada termo
  print(token.text, token.pos_, token.dep_) # exibe o texto a classe gramatical e a função sintática
```

```python
from spacy import displacy # importa displacy de ferramenta

displacy.render(doc2, style = "dep") # imprime a arvore de dependencias
```
