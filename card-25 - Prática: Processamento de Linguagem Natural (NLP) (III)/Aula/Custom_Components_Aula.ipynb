```python
import spacy # importa a biblioteca spacy
```

```python
nlp = spacy.load("en_core_web_sm") # carrega o modelo de linguagem
doc = nlp("Britain is a place. Mary is as doctor.") # processa o texto inicial
```

```python
for ent in doc.ents: # percorre as entidades encontrada
    print(ent.text, ent.label_) # exibe o texto e o rótulo da entidade
```

```python
from spacy.language import Language
```

```python
@Language.component("remove_gpe") # define um componente
def remove_gpe(doc): # função que processa o documento
    original_ents = list(doc.ents) # converte as entidades para uma lista
    for ent in doc.ents: # itera sobre cada entidade detectada
        if ent.label_ == "GPE": # verifica se a etiqueta é de localidade
            original_ents.remove(ent) # remove a entidade da lista temporária
    doc.ents = original_ents # sobrescreve as entidades do documento
    return doc # retorna o documento alterado
```

```python
nlp.add_pipe("remove_gpe") # adiciona a função ao pipeline
```

```python
nlp.analyze_pipes() # exibe a estrutura atual do pipeline
```

```python
doc = nlp("Britain is a place. Mary is as doctor.")
for ent in doc.ents: # percorre as entidades restantes
    print(ent.text, ent.label_) # exibe os resultados filtrados
```
