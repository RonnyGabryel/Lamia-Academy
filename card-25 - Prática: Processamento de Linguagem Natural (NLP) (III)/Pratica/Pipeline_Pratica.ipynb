```python
# criando um pipeline vazio em inglês e adicionando apenas o sentencizer
nlp_blank = spacy.blank("en")
nlp_blank.add_pipe("sentencizer")

nlp_blank.analyze_pipes() # inspecionando o pipeline minimalista
```

```python
# carregando o modelo medio completo para comparação
nlp_full = spacy.load("en_core_web_md")

nlp_full.analyze_pipes() # inspecionando todos os componentes do pipeline completo
```

```python
with open("historico_wow.txt", "r") as f: # abrindo o arquivo do WoW
    text = f.read()
```

```python
doc_blank = nlp_blank(text) # processando o texto com o pipeline

for sent in list(doc_blank.sents)[:5]: # exibindo as cinco primeiras sentenças detectadas
    print(sent)
```

```python
# processando o mesmo texto com o pipeline completo
doc_full = nlp_full(text)
```

```python
# o pipeline minimalista não tem NER, então doc.ents estará vazio
print("pipeline vazio:", list(doc_blank.ents))

# o pipeline completo reconhece entidades normalmente
print("pipeline completo:", [(ent.text, ent.label_) for ent in doc_full.ents[:10]])
```

```python
# o pipeline não tem tagger
token_blank = list(doc_blank)[5]
print("pipeline vazio:", token_blank.pos_)

# pipeline com o tagger preenchido
token_full = list(doc_full)[5]
print("POS no pipeline completo:", token_full.pos_)
```
