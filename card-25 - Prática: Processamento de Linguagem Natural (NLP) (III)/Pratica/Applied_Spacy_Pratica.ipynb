```python
# importando tudo necessário para a aula
import spacy
import pandas as pd
from spacy import displacy
```

```python
# montando os dados estruturados do universo WoW em dicionários

dados_wow = {
    "classes": [
        "Death Knight", "Druid", "Hunter", "Paladin",
        "Priest", "Rogue", "Shaman", "Warlock", "Warrior"
    ],
    "racas": [
        "Human", "Orc", "Undead", "Tauren", "Night Elf",
        "Dwarf", "Gnome", "Troll", "Blood Elf", "Draenei"
    ],
    "habilidades": [
        "Bloodlust", "Aura Mastery", "Hammer of Justice", "Holy Wrath",
        "Anti-Magic Shell", "Divine Shield", "Pain Suppression",
        "Divine Hymn", "Righteous Fury", "Ardent Defender",
        "Tricks of the Trade", "Misdirection", "Kidney Shot",
        "Crippling Poison", "Desecration", "Feign Death",
        "Fade", "Starfall", "Typhoon"
    ],
    "jogadores": [
        "Lazeil", "Sejta", "Lappé", "Synti",
        "Jhazrun", "Ilonie", "Diamondtear"
    ],
    "guilds": [
        "Paragon"
    ],
    "bosses": [
        "Lich King", "Val'kyr Shadowguard", "Shambling Horrors",
        "Raging Spirits", "Vile Spirits", "Frozen Orbs"
    ],
    "locais": [
        "Icecrown Citadel", "Frostmourne Room"
    ],
    "itens": [
        "Frostmourne"
    ]
}
```

```python
# conferindo o conteúdo de cada categoria antes de gerar os padrões
for categoria, itens in dados_wow.items():
    print(categoria, len(itens), "itens")
```

```python
# termos que podem gerar falso positivo e precisam ser ignorados
stops = ["two", "three", "Holy", "Shadow", "Combat", "Fury", "Balance"]

nlp = spacy.blank("en")
ruler = nlp.add_pipe("entity_ruler")

patterns = [] # lista que vai receber todos os padrões gerados dinamicamente

for classe in dados_wow["classes"]:
    if classe not in stops:
        patterns.append({"label": "CLASS", "pattern": classe})

for raca in dados_wow["racas"]:
    if raca not in stops:
        patterns.append({"label": "RACE", "pattern": raca})

for habilidade in dados_wow["habilidades"]:
    if habilidade not in stops:
        patterns.append({"label": "ABILITY", "pattern": habilidade})

for jogador in dados_wow["jogadores"]:
    patterns.append({"label": "PLAYER", "pattern": jogador})

for guild in dados_wow["guilds"]:
    patterns.append({"label": "GUILD", "pattern": guild})

for boss in dados_wow["bosses"]:
    patterns.append({"label": "BOSS", "pattern": boss})

for local in dados_wow["locais"]:
    patterns.append({"label": "LOCATION", "pattern": local})

for item in dados_wow["itens"]:
    patterns.append({"label": "ITEM", "pattern": item})

ruler.add_patterns(patterns) # injetando todos os padrões gerados no ruler

print(len(patterns))# padrões carregados no ruler
```

```python
with open("historico_wow.txt", "r") as f:
    text = f.read()

doc = nlp(text)

for ent in doc.ents: # exibindo todas as entidades
    print(ent.text, ent.label_)
```
