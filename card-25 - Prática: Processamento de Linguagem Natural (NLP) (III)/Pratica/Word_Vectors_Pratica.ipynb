```python
import spacy # importando a biblioteca spacy
import numpy as np # importando numpy para operações com vetores
!python -m spacy download en_core_web_md # baixando o modelo médio que contém vetores de palavras
nlp = spacy.load("en_core_web_md") # carregando o modelo médio
```

```python
with open("historico_wow.txt", "r") as f: # abrindo o arquivo do wow
    text = f.read()

doc = nlp(text) # processando o texto pelo pipeline do spacy
sentence1 = list(doc.sents)[0] # pegando a primeira sentença
print(sentence1)
```

```python
your_word = "paladin" # palavra do universo no WoW

ms = nlp.vocab.vectors.most_similar(
    np.asarray([nlp.vocab.vectors[nlp.vocab.strings[your_word]]]), n=10)

words = [nlp.vocab.strings[w] for w in ms[0][0]] # convertendo ids para palavras
distances = ms[2] # guardando as distâncias

print(words) # exibindo as palavras
```

```python
# comparando duas frases sobre estratégia de raid
doc1 = nlp("Lazeil tanked the Shambling Horrors using stuns and cooldowns.")
doc2 = nlp("Sejta held the Lich King while the raid focused on adds.")

print(doc1, "<->", doc2, doc1.similarity(doc2))
```

```python
# comparando uma frase de raid com algo completamente fora do contexto
doc3 = nlp("The weather in London is cold and rainy in November.")

print(doc1, "<->", doc3, doc1.similarity(doc3))
```

```python
# comparando frases sobre habilidades parecidas entre classes diferentes
doc4 = nlp("The Paladin used Hammer of Justice to stun the Val'kyr.")
doc5 = nlp("The Rogue used Kidney Shot to stun the Val'kyr.")

print(doc4, "<->", doc5, doc4.similarity(doc5))
```

```python
# comparando habilidades de papel completamente diferente
doc6 = nlp("Jhazrun shielded the raid with Power Word Shield before Infest.")

print(doc4, "<->", doc6, doc4.similarity(doc6))
```
