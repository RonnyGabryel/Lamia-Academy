```python
!pip install spacy # instalando o spacy
!python -m spacy download en_core_web_sm # baixando o modelo de inglês
import spacy # carregando a biblioteca
nlp = spacy.load("en_core_web_sm") # confirmando que o modelo foi carregado
from spacy import displacy # importando displacy para visualização
```

```python
with open("historico_wow.txt", "r") as f: # abrindo o arquivo
    text = f.read()
```

```python
doc = nlp(text) # processando o texto
```

```python
print(len(text)) # contando os caracteres do texto bruto
print(len(doc))  # contando os tokens processados pelo spacy
```

```python
for token in text[:10]: # percorrendo os dez primeiros símbolos do texto bruto
  print(token)
```

```python
for token in doc[:10]: # percorrendo os dez primeiros tokens
  print(token)
```

```python
sentence1 = list(doc.sents)[0] # convertendo o gerador em lista
print(sentence1)
```

```python
token2 = sentence1[2] # acessando o terceiro token da primeira sentença
print(token2.text) # texto do token
print(token2.lemma_) # forma base não flexionada
print(token2.pos_) # classe gramatical
print(token2.dep_)  # função sintática
print(token2.morph) # características morfológicas
print(token2.ent_iob_)# posição em relação à entidade
print(token2.lang_) # idioma identificado
```

```python
text2 = "Paragon defeated the Lich King on Icecrown Citadel using Bloodlust during the Frostmourne Room phase." # frase com múltiplas entidades do universo WoW
doc2 = nlp(text2) # processando pelo pipeline
```

```python
for token in doc2: # percorrendo cada token
    print(token.text, token.pos_, token.dep_) # exibindo texto
```

```python
displacy.render(doc2, style="dep") # renderizando a árvore de dependências
```

```python
displacy.render(doc2, style="ent") # visualizando as entidades
```

```python
from collections import Counter # importando counter para contagem

contagem = Counter(ent.label_ for ent in doc.ents) # contando quantas entidades de cada tipo aparecem no texto

for tipo, quantidade in contagem.most_common(): # exibindo do tipo mais frequente ao menos frequente
    print(tipo, quantidade)
```

```python
displacy.render(doc, style="ent") # visualização geral de todas as entidades do texto
```
