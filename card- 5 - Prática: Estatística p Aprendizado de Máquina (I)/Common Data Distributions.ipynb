```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Distribuição Uniforme
print("\n1. Distribuição Uniforme")
print("Todos os valores têm a mesma chance de ocorrer")

valores = np.random.uniform(-10.0, 10.0, 100000)
plt.hist(valores,50)
plt.show()
```

```python
# Distribuição Normal - PDF teórica
print("\n2. Distribuição Normal (Curva Teórica)")
print("A famosa curva em formato de sino")

from scipy.stats import norm

x = np.arange(-3, 3, 0.001)
plt.plot(x, norm.pdf(x))
```

```python
# Distribuição Normal - Amostragem
print("\n3. Distribuição Normal (Amostra)")
print("Gerando dados que seguem a distribuição normal")

mu = 5.0
sigma = 2.0
valores = np.random.normal(mu, sigma, 10000)
plt.hist(valores, 50)
plt.show()
```

```python
# Distribuição Exponencial
print("\n4. Distribuição Exponencial")
print("Útil para modelar tempo entre eventos")

from scipy.stats import expon

x = np.arange(0, 10, 0.001)
plt.plot(x, expon.pdf(x))
```

```python
# Distribuição Binomial
print("\n5. Distribuição Binomial")
print("Número de sucessos em tentativas independentes")
from scipy.stats import binom

n,p = 10, 0.5
x = np.arange(0, 10, 0.001)
plt.plot(x, binom.pmf(x, n, p))
```

```python
# Distribuição de Poisson
print("\n6. Distribuição de Poisson")
print("Número de eventos em intervalo fixo de tempo")
from scipy.stats import poisson

mu = 500
x = np.arange(400, 600, 0.5)
plt.plot(x, poisson.pmf(x, mu))
```
