```python
# Importa o NumPy para manipulação eficiente de arrays e operações matemáticas.
import numpy as np

# Importa o Matplotlib para criação de gráficos e visualizações personalizadas.
import matplotlib.pyplot as plt
```

```python
# Gera um vetor de 10.000 números aleatórios seguindo uma distribuição normal
# com média 100 e desvio padrão 20.
vetor = np.random.normal(100.0, 20.0, 10000)

# Plota um histograma com 50 bins para visualizar a distribuição dos dados.
plt.hist(vetor, 50)
plt.show()
```

```python
vetor.std() #desvio padrão
```

```python
vetor.var() #variancia
```
