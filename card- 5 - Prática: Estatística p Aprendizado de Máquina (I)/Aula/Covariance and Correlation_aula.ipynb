```python
# Aqui estão sendo importadas as ferramentas necessárias:
# NumPy para cálculos matemáticos, Matplotlib para criar gráficos
# e o módulo random do NumPy para gerar dados aleatórios.
import numpy as np
import matplotlib.pyplot as plt
from numpy import dot
```

```python
# Função para calcular a média de um conjunto de valores.
def de_mean(x):
    xmean=np.mean(x)
    return[xi - xmean for xi in x]

# Função para calcular a covariância entre dois conjuntos de dados.
# A covariância mede como duas variáveis variam juntas.
def covariance(x, y):
    n=len(x)
    return dot (de_mean(x), de_mean(y))/ (n-1)

# Geração de dois conjuntos de dados aleatórios:
# pageSpeeds representa a velocidade de carregamento de páginas (distribuição normal com média 3 e desvio padrão 1).
# purchaseAmount representa o valor gasto em compras (distribuição normal com média 50 e desvio padrão 10).
pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = np.random.normal(50.0, 10.0, 1000)
```

```python
# Gráfico de dispersão entre velocidade da página e valor de compra.
plt.scatter(pageSpeeds, purchaseAmount)
# Impressão da covariância entre as duas variáveis.
print("Covariância:", covariance(pageSpeeds, purchaseAmount))
```

```python
# Alterando os dados de purchaseAmount para que dependam da velocidade da página.
# Agora, quanto maior o pageSpeeds, menor tende a ser o purchaseAmount.
purchaseAmount=np.random.normal(50.0, 10.0, 1000)/pageSpeeds
# Gráfico de dispersão atualizado.
plt.scatter(pageSpeeds, purchaseAmount)
# Impressão da nova covariância.
print("Covariância:", covariance(pageSpeeds, purchaseAmount))
```

```python
# Função para calcular a correlação entre duas variáveis.
# A correlação é uma versão "padronizada" da covariância,
# dividida pelo desvio padrão de cada variável.
def correlation(x,y):
    stddevx=x.std()
    stddevy=y.std()
    return covariance(x,y)/ stddevx/ stddevy
correlation(pageSpeeds, purchaseAmount)
```

```python
# Impressão da correlação entre pageSpeeds e purchaseAmount.
np.corrcoef(pageSpeeds, purchaseAmount)
```

```python
# Ajustando novamente os dados de purchaseAmount para simular outro cenário.
purchaseAmount=100-pageSpeeds*3
# Gráfico de dispersão entre pageSpeeds e purchaseAmount.
plt.scatter(pageSpeeds, purchaseAmount)
# Impressão da covariância com a nova relação entre as variáveis.
print("Covariância:", covariance(pageSpeeds, purchaseAmount))
```
