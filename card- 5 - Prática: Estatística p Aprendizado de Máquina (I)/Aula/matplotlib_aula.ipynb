```python
# Importa a biblioteca 'matplotlib.pyplot' para criação de gráficos 
# e a referencia como 'plt' para facilitar o uso
import matplotlib.pyplot as plt  
# Define a fonte padrão dos gráficos do matplotlib para "DejaVu Sans" 
plt.rcParams["font.family"] = "DejaVu Sans"  

# Importa a função 'norm' da biblioteca scipy.stats, utilizada para
# trabalhar com distribuições normais (gaussianas)
from scipy.stats import norm  

# Importa novamente matplotlib.pyplot e numpy, garantindo que 
# 'plt' e 'np' estejam disponíveis para criação de gráficos e manipulação de arrays
import matplotlib.pyplot as plt  
import numpy as np  
```

```python
#Cria um arrar com valores de -3 a 3 com intervalos de '0,001'
x = np.arange(-3,3,0.001)
```

```python
#Plota o gráfico do tipo histograma da função FDP do conjunto dos dados considerando uma distribuição normal padrão
plt.plot(x,norm.pdf(x))
plt.show()
```

```python
# lota o gráfico do tipo histograma da função FDP do conjunto dos dados considerando uma distribuição normal padrão, com mediana '0' e desvio padrão '1'
plt.plot(x,norm.pdf(x))
plt.plot(x,norm.pdf(x,1.0,0.5))
plt.show()
```

```python
# Cria uma lista com os valores que serão representados em cada fatia da pizza
Valores = [12, 55, 4, 32, 14]  

# Cria uma lista de cores para cada fatia do gráfico
Cores = ['r', 'g', 'b', 'c', 'm']  

# Cria uma lista indicando quais fatias serão destacadas (explodidas) no gráfico
# Um valor maior que 0 "afasta" a fatia do centro da pizza
Explosao = [0, 0, 0.2, 0, 0]  

# Cria uma lista com os nomes das categorias correspondentes a cada fatia
Eixos = ['India', 'Estados Unidos', 'Russia', 'China', 'Europa']  

# Plota o gráfico de pizza (pie chart) usando os valores, cores, fatias explodidas e rótulos
plt.pie(Valores, colors=Cores, explode=Explosao, labels=Eixos)  

# Adiciona um título ao gráfico
plt.title('Localidade Dos Estudantes')  

# Exibe o gráfico de pizza final
plt.show()  
```

```python
#Plota o gráfico de barras, indo de 0 a 5 no eixo x, utilizando os parâmetros inseridos nas listas anteriormente
plt.bar(range(0,5), Valores, color = Cores)
plt.show()
```

```python
#Cria um array de 10000 valores aleatórios normalmente distribuidos com mediana de '27000'e desvio padrão '15000' e divide em 50 classes
Rendas = np.random.normal(27000, 15000, 10000)
plt.hist(Rendas, 50)
plt.show()
```

```python
# Gera 100 valores aleatórios seguindo uma distribuição uniforme, 
# multiplicando por 100 para aumentar a escala e subtraindo 40 para deslocar o intervalo. 
Assimetria_uniforme = np.random.rand(100) * 100 - 40  

# Gera 10 valores aleatórios que serão outliers superiores (valores altos),
# multiplicando por 50 e somando 100 para garantir que fiquem acima da maioria dos dados.  
Outliers_Superiores = np.random.rand(10) * 50 + 100  

# Gera 10 valores aleatórios que serão outliers inferiores (valores muito baixos),
# multiplicando por -50 e subtraindo 100 para garantir que fiquem abaixo da maioria dos dados.  
Outliers_Inferiores = np.random.rand(10) * -50 - 100  

# Concatena os três conjuntos de dados (valores principais e outliers)
# em um único array chamado 'Dados'.  
Dados = np.concatenate((Assimetria_uniforme, Outliers_Superiores, Outliers_Inferiores))  

# Cria um boxplot (gráfico de caixa) para visualizar a distribuição dos dados,
# incluindo mediana, quartis e possíveis outliers.  
plt.boxplot(Dados)  

# Exibe o gráfico na tela.  
plt.show() 
```
