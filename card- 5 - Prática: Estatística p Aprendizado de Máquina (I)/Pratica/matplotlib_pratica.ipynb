```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, t
import pandas as pd

# Configurações iniciais
np.random.seed(42)
```

```python
# Primeiro gráfico - Distribuição Normal
x = np.linspace(-4, 4, 100)
y_normal = norm.pdf(x, 0, 1)

plt.figure(figsize=(8, 5))
plt.plot(x, y_normal, 'b-', linewidth=2, label='Normal(0,1)')
plt.xlabel('X')
plt.ylabel('Densidade')
plt.title('Distribuição Normal Padrão')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

```python
# Segundo gráfico - Comparando Normal e t-Student
x = np.linspace(-4, 4, 100)
y_normal = norm.pdf(x, 0, 1)
y_t = t.pdf(x, df=5)

plt.figure(figsize=(8, 5))
plt.plot(x, y_normal, 'b-', linewidth=2, label='Normal')
plt.plot(x, y_t, 'r-', linewidth=2, label='t-Student (df=5)')
plt.xlabel('X')
plt.ylabel('Densidade')
plt.title('Comparação: Normal vs t-Student')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Terceiro gráfico - Diferentes graus de liberdade
x = np.linspace(-4, 4, 100)
y_normal = norm.pdf(x, 0, 1)
y_t3 = t.pdf(x, df=3)
y_t10 = t.pdf(x, df=10)

plt.figure(figsize=(8, 5))
plt.plot(x, y_normal, 'b-', linewidth=2, label='Normal')
plt.plot(x, y_t3, 'r-', linewidth=2, label='t (df=3)')
plt.plot(x, y_t10, 'g-', linewidth=2, label='t (df=10)')
plt.xlabel('X')
plt.ylabel('Densidade')
plt.title('t-Student com Diferentes GL')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Gerando dados para análise
amostra_normal = np.random.normal(0, 1, 1000)
amostra_t = np.random.standard_t(5, 1000)
```

```python
# Criando DataFrame com os dados
dados = pd.DataFrame({
    'indice': range(1000),
    'normal': amostra_normal,
    't_student': amostra_t,
    'diferenca': amostra_normal - amostra_t
})

print("\nPrimeiras 50 observações:")
print(dados.head(50))
```

```python
# Gráfico adicional após a tabela
x = np.linspace(-4, 4, 100)
y_normal = norm.pdf(x, 0, 1)
y_t = t.pdf(x, df=8)

plt.figure(figsize=(8, 5))
plt.plot(x, y_normal, 'b-', linewidth=2, label='Normal')
plt.plot(x, y_t, 'orange', linewidth=2, label='t-Student (df=8)')
plt.xlabel('X')
plt.ylabel('Densidade')
plt.title('Comparação Final das Distribuições')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Segunda tabela de dados
print("\nSegunda amostra de dados:")
amostra2_normal = np.random.normal(1, 1.5, 500)
amostra2_t = np.random.standard_t(3, 500)

dados2 = pd.DataFrame({
    'indice': range(500),
    'normal2': amostra2_normal,
    't_student2': amostra2_t,
    'produto': amostra2_normal * amostra2_t
})

for i in range(500):
    print(f"{i+1:4d} {dados2.iloc[i]['normal2']:10.6f} {dados2.iloc[i]['t_student2']:10.6f} {dados2.iloc[i]['produto']:10.6f}")
```

```python
# Estatísticas finais
print(f"\nEstatísticas finais:")
print(f"Média amostra 1 (normal): {dados['normal'].mean():.6f}")
print(f"Média amostra 1 (t-student): {dados['t_student'].mean():.6f}")
print(f"Média amostra 2 (normal): {dados2['normal2'].mean():.6f}")
print(f"Média amostra 2 (t-student): {dados2['t_student2'].mean():.6f}")

print(f"\nDesvios padrão:")
print(f"Desvio amostra 1 (normal): {dados['normal'].std():.6f}")
print(f"Desvio amostra 1 (t-student): {dados['t_student'].std():.6f}")
print(f"Desvio amostra 2 (normal): {dados2['normal2'].std():.6f}")
print(f"Desvio amostra 2 (t-student): {dados2['t_student2'].std():.6f}")

print(f"\nAnálise completa das distribuições finalizada!")
print(f"Total de observações processadas: {len(dados) + len(dados2)}")
```

```python
# Terceira tabela - Dados combinados
print(f"\nTerceira amostra - Dados combinados:")
amostra3_x = np.random.normal(0, 1, 300)
amostra3_y = np.random.exponential(1, 300)

dados3 = pd.DataFrame({
    'indice': range(300),
    'x_normal': amostra3_x,
    'y_exponencial': amostra3_y,
    'razao': amostra3_x / (amostra3_y + 0.001)  # evita divisão por zero
})

for i in range(300):
    print(f"{i+1:4d} {dados3.iloc[i]['x_normal']:10.6f} {dados3.iloc[i]['y_exponencial']:10.6f} {dados3.iloc[i]['razao']:10.6f}")

print(f"\nResumo final:")
print(f"Amostra 1: {len(dados)} observações")
print(f"Amostra 2: {len(dados2)} observações") 
print(f"Amostra 3: {len(dados3)} observações")
print(f"Total geral: {len(dados) + len(dados2) + len(dados3)} observações")
```

```python
# Gráfico final com legenda
plt.figure(figsize=(8, 6))
x = np.linspace(-4, 4, 100)
y1 = norm.pdf(x, 0, 1)
y2 = t.pdf(x, df=5)

plt.plot(x, y1, 'b-', linewidth=2, label='Var 1')
plt.plot(x, y2, 'orange', linewidth=2, label='Var 2') 
plt.xlabel('X')
plt.ylabel('Densidade')
plt.title('Comparação Final')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Última tabela de dados
print(f"\nÚltima amostra de dados:")
amostra_final = np.random.gamma(2, 2, 400)
log_amostra = np.log(amostra_final + 1)

dados_finais = pd.DataFrame({
    'indice': range(400),
    'gamma': amostra_final,
    'log_gamma': log_amostra,
    'quadrado': amostra_final ** 2
})

for i in range(400):
    print(f"{i+1:4d} {dados_finais.iloc[i]['gamma']:10.6f} {dados_finais.iloc[i]['log_gamma']:10.6f} {dados_finais.iloc[i]['quadrado']:12.6f}")
```

```python
# Gráfico de dispersão final
plt.figure(figsize=(8, 6))
x_scatter = np.random.normal(0, 2, 500)
y_scatter = x_scatter + np.random.normal(0, 1, 500)

plt.scatter(x_scatter, y_scatter, alpha=0.6, s=15, color='blue')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gráfico de Dispersão')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Histograma azul
dados_hist = np.random.normal(100, 15, 2000)
plt.figure(figsize=(8, 6))
plt.hist(dados_hist, bins=30, color='skyblue', alpha=0.8, edgecolor='blue')
plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.title('Distribuição dos Dados')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Estatísticas do histograma
print(f"Estatísticas do histograma:")
print(f"Média: {dados_hist.mean():.6f}")
print(f"Mediana: {np.median(dados_hist):.6f}")
print(f"Desvio padrão: {dados_hist.std():.6f}")
print(f"Mínimo: {dados_hist.min():.6f}")
print(f"Máximo: {dados_hist.max():.6f}")
```

```python
# Nova tabela de dados
print(f"\nNova tabela de dados experimentais:")
experimento = np.random.lognormal(2, 0.8, 600)
controle = np.random.normal(50, 12, 600)

dados_exp = pd.DataFrame({
    'id': range(600),
    'experimento': experimento,
    'controle': controle,
    'diferenca': experimento - controle
})

for i in range(600):
    print(f"{i+1:4d} {dados_exp.iloc[i]['experimento']:12.6f} {dados_exp.iloc[i]['controle']:10.6f} {dados_exp.iloc[i]['diferenca']:12.6f}")
```
