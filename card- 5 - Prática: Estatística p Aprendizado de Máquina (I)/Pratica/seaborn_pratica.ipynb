```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Configurando o estilo dos gr√°ficos
plt.style.use('seaborn-v0_8')
np.random.seed(42)
```

```python
# Criando um dataset exemplo
data = {
    'vendas': np.random.exponential(2, 1000),
    'idade': np.random.normal(35, 10, 1000),
    'salario': np.random.lognormal(10, 0.5, 1000),
    'experiencia': np.random.gamma(2, 2, 1000)
}
df = pd.DataFrame(data)
```

```python
# Primeiro gr√°fico - Distribui√ß√£o das vendas
plt.figure(figsize=(8, 5))
plt.hist(df['vendas'], bins=20, color='skyblue', alpha=0.7)
plt.title('Distribui√ß√£o das Vendas')
plt.xlabel('Vendas')
plt.ylabel('Frequ√™ncia')
plt.show()
```

```python
# Segundo gr√°fico - Distribui√ß√£o da idade
plt.figure(figsize=(8, 5))
plt.hist(df['idade'], bins=25, color='lightblue', alpha=0.7)
plt.title('Distribui√ß√£o da Idade')
plt.xlabel('Idade')
plt.ylabel('Frequ√™ncia')
plt.show()
```

```python
# Estat√≠sticas b√°sicas
print("Estat√≠sticas Descritivas:")
print(df.describe())
```

```python
# Terceiro gr√°fico - Histograma do sal√°rio
plt.figure(figsize=(8, 5))
plt.hist(df['salario'], bins=30, color='steelblue', alpha=0.7)
plt.title('Distribui√ß√£o do Sal√°rio')
plt.xlabel('Sal√°rio')
plt.ylabel('Frequ√™ncia')
plt.show()
```

```python
# Matriz de correla√ß√£o
print("\nMatriz de Correla√ß√£o:")
correlation_matrix = df.corr()
print(correlation_matrix)
```

```python
# Quarto gr√°fico - Pairplot para visualizar rela√ß√µes
plt.figure(figsize=(12, 10))
sns.pairplot(df, diag_kind='hist', plot_kws={'alpha':0.6})
plt.suptitle('Matriz de Dispers√£o - Todas as Vari√°veis', y=1.02)
plt.show()
```

```python
# Gr√°ficos de dispers√£o individuais
# Vendas vs Idade
plt.figure(figsize=(8, 6))
plt.scatter(df['idade'], df['vendas'], alpha=0.6, s=20)
plt.xlabel('Idade')
plt.ylabel('Vendas')
plt.title('Rela√ß√£o entre Idade e Vendas')
plt.show()
```

```python
# Vendas vs Sal√°rio
plt.figure(figsize=(8, 6))
plt.scatter(df['salario'], df['vendas'], alpha=0.6, s=20)
plt.xlabel('Sal√°rio')
plt.ylabel('Vendas')
plt.title('Rela√ß√£o entre Sal√°rio e Vendas')
plt.show()
```

```python
# Experi√™ncia vs Vendas
plt.figure(figsize=(8, 6))
plt.scatter(df['experiencia'], df['vendas'], alpha=0.6, s=20)
plt.xlabel('Experi√™ncia')
plt.ylabel('Vendas')
plt.title('Rela√ß√£o entre Experi√™ncia e Vendas')
plt.show()
```

```python
# Criando dados categ√≥ricos para an√°lise
np.random.seed(123)
categorias = ['Janeiro', 'Fevereiro', 'Mar√ßo', 'Abril', 'Maio', 'Junho', 
              'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']

vendas_mensais = np.random.uniform(50, 200, 12)
lucro_mensal = np.random.uniform(20, 80, 12)

# Criando DataFrame mensal
df_mensal = pd.DataFrame({
    'Mes': categorias,
    'Vendas': vendas_mensais,
    'Lucro': lucro_mensal
})

print("Dados Mensais:")
print(df_mensal)
```

```python
# Gr√°fico de barras - Vendas mensais
plt.figure(figsize=(12, 6))
plt.bar(df_mensal['Mes'], df_mensal['Vendas'], alpha=0.7, color='steelblue')
plt.xlabel('M√™s')
plt.ylabel('Vendas')
plt.title('Vendas Mensais')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Gr√°fico de barras - Lucro mensal
plt.figure(figsize=(12, 6))
plt.bar(df_mensal['Mes'], df_mensal['Lucro'], alpha=0.7, color='lightcoral')
plt.xlabel('M√™s')
plt.ylabel('Lucro')
plt.title('Lucro Mensal')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
