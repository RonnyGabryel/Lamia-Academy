```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Configurando o estilo dos gráficos
plt.style.use('seaborn-v0_8')
np.random.seed(42)
```

```python
# Criando um dataset exemplo
data = {
    'vendas': np.random.exponential(2, 1000),
    'idade': np.random.normal(35, 10, 1000),
    'salario': np.random.lognormal(10, 0.5, 1000),
    'experiencia': np.random.gamma(2, 2, 1000)
}
df = pd.DataFrame(data)
```

```python
# Primeiro gráfico - Distribuição das vendas
plt.figure(figsize=(8, 5))
plt.hist(df['vendas'], bins=20, color='skyblue', alpha=0.7)
plt.title('Distribuição das Vendas')
plt.xlabel('Vendas')
plt.ylabel('Frequência')
plt.show()
```

```python
# Segundo gráfico - Distribuição da idade
plt.figure(figsize=(8, 5))
plt.hist(df['idade'], bins=25, color='lightblue', alpha=0.7)
plt.title('Distribuição da Idade')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
```

```python
# Estatísticas básicas
print("Estatísticas Descritivas:")
print(df.describe())
```

```python
# Terceiro gráfico - Histograma do salário
plt.figure(figsize=(8, 5))
plt.hist(df['salario'], bins=30, color='steelblue', alpha=0.7)
plt.title('Distribuição do Salário')
plt.xlabel('Salário')
plt.ylabel('Frequência')
plt.show()
```

```python
# Matriz de correlação
print("\nMatriz de Correlação:")
correlation_matrix = df.corr()
print(correlation_matrix)
```

```python
# Quarto gráfico - Pairplot para visualizar relações
plt.figure(figsize=(12, 10))
sns.pairplot(df, diag_kind='hist', plot_kws={'alpha':0.6})
plt.suptitle('Matriz de Dispersão - Todas as Variáveis', y=1.02)
plt.show()
```

```python
# Gráficos de dispersão individuais
# Vendas vs Idade
plt.figure(figsize=(8, 6))
plt.scatter(df['idade'], df['vendas'], alpha=0.6, s=20)
plt.xlabel('Idade')
plt.ylabel('Vendas')
plt.title('Relação entre Idade e Vendas')
plt.show()
```

```python
# Vendas vs Salário
plt.figure(figsize=(8, 6))
plt.scatter(df['salario'], df['vendas'], alpha=0.6, s=20)
plt.xlabel('Salário')
plt.ylabel('Vendas')
plt.title('Relação entre Salário e Vendas')
plt.show()
```

```python
# Experiência vs Vendas
plt.figure(figsize=(8, 6))
plt.scatter(df['experiencia'], df['vendas'], alpha=0.6, s=20)
plt.xlabel('Experiência')
plt.ylabel('Vendas')
plt.title('Relação entre Experiência e Vendas')
plt.show()
```

```python
# Criando dados categóricos para análise
np.random.seed(123)
categorias = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 
              'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']

vendas_mensais = np.random.uniform(50, 200, 12)
lucro_mensal = np.random.uniform(20, 80, 12)

# Criando DataFrame mensal
df_mensal = pd.DataFrame({
    'Mes': categorias,
    'Vendas': vendas_mensais,
    'Lucro': lucro_mensal
})

print("Dados Mensais:")
print(df_mensal)
```

```python
# Gráfico de barras - Vendas mensais
plt.figure(figsize=(12, 6))
plt.bar(df_mensal['Mes'], df_mensal['Vendas'], alpha=0.7, color='steelblue')
plt.xlabel('Mês')
plt.ylabel('Vendas')
plt.title('Vendas Mensais')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Gráfico de barras - Lucro mensal
plt.figure(figsize=(12, 6))
plt.bar(df_mensal['Mes'], df_mensal['Lucro'], alpha=0.7, color='lightcoral')
plt.xlabel('Mês')
plt.ylabel('Lucro')
plt.title('Lucro Mensal')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
