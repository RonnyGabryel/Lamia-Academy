```python
# Importando bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.stats import binom
from scipy.stats import poisson
```

```python
# Gerando dados uniformes
np.random.seed(42)
valores = np.random.uniform(-10, 10, 500)
```

```python
# Estatísticas básicas
print("Média:", valores.mean())
print("Desvio padrão:", valores.std())
```

```python
# Histograma dos dados uniformes
plt.figure(figsize=(8, 5))
plt.hist(valores, bins=20, color='skyblue', alpha=0.7)
plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.title('Distribuição Uniforme')
plt.show()
```

```python
# Importando scipy.stats
from scipy.stats import norm
```

```python
# Gerando curva normal padrão para comparação
x = np.linspace(-3, 3, 100)
y = norm.pdf(x, 0, 1)
```

```python
# Plotando curva normal
plt.figure(figsize=(8, 5))
plt.plot(x, y, color='blue', linewidth=2)
plt.xlabel('Z')
plt.ylabel('Densidade')
plt.title('Distribuição Normal Padrão')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Gerando nova amostra normal
mu = 5.0
sigma = 2.0
```

```python
# Nova amostra com distribuição normal
valores2 = np.random.normal(mu, sigma, 1000)
print(f"\nNova amostra:")
print(f"Mu = {mu}")
print(f"Sigma = {sigma}")
```

```python
# Histograma da nova amostra
plt.figure(figsize=(8, 5))
plt.hist(valores2, bins=25, color='skyblue', alpha=0.7)
plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.title('Nova Distribuição Normal')
plt.show()
```

```python
# Parâmetro de Poisson
mu_poisson = 3.0
x_poisson = np.arange(0, 12)
y_poisson = poisson.pmf(x_poisson, mu_poisson)

plt.figure(figsize=(8, 5))
plt.bar(x_poisson, y_poisson, color='skyblue', alpha=0.7)
plt.xlabel('Número de Eventos')
plt.ylabel('Probabilidade')
plt.title('Distribuição de Poisson')
plt.show()
```
