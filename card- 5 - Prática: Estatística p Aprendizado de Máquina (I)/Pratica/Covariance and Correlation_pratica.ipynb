```python
# Biblioteca basicas
import numpy as np
import matplotlib.pyplot as plt

# Configurando seed
np.random.seed(45)
```

```python
# Primeiro grfico - dados aleatrios
n = 500
x1 = np.random.randn(n)
y1 = np.random.randn(n)

plt.figure(figsize=(8, 6))
plt.scatter(x1, y1, alpha=0.6, s=20)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Dados Aleatrios')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Segundo grfico - relao exponencial
x2 = np.linspace(0, 5, 200)
y2 = np.exp(-x2) + np.random.normal(0, 0.1, 200)

plt.figure(figsize=(8, 6))
plt.scatter(x2, y2, alpha=0.7, s=15)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Relao Exponencial')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Terceiro grfico - correlao linear negativa
x3 = np.linspace(0, 10, 100)
y3 = -1.5 * x3 + 15 + np.random.normal(0, 2, 100)

plt.figure(figsize=(8, 6))
plt.scatter(x3, y3, alpha=0.7, s=25)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Correlao Negativa')
plt.grid(True, alpha=0.3)
```

```python
# Adicionando linha de tendncia
z = np.polyfit(x3, y3, 1)
p = np.poly1d(z)
plt.plot(x3, p(x3), "r--", alpha=0.8)

plt.show()
```
