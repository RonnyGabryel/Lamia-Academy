```python
# Importando as bibliotecas básicas
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Gerando uma amostra de dados aleatórios
np.random.seed(123)
dados = np.random.normal(100, 15, 1000)
```

```python
# Calculando estatísticas básicas
print("Média:", dados.mean())
print("Desvio padrão:", dados.std())
print("Tamanho da amostra:", len(dados))
```

```python
# Fazendo o histograma
plt.figure(figsize=(8, 6))
plt.hist(dados, bins=30, alpha=0.7, color='skyblue')
plt.title('Histograma dos Dados')
plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.show()
```

```python
# Algumas estatísticas adicionais
print("\nMínimo:", dados.min())
print("Máximo:", dados.max())
print("Mediana:", np.median(dados))
```

```python
# Segundo conjunto de dados para comparação
dados2 = np.random.normal(110, 20, 800)
print("Média:", dados2.mean())
print("Desvio padrão:", dados2.std())
print("Tamanho da amostra:", len(dados2))
```

```python
# Histograma do segundo conjunto
plt.figure(figsize=(8, 6))
plt.hist(dados2, bins=25, alpha=0.7, color='skyblue')
plt.title('Segundo Histograma')
plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.show()
```
