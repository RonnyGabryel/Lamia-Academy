```python
import torch # importação da biblioteca principal pytorch
from torchviz import make_dot # importa a ferramenta de visualização de grafos
```

```python
x = [[1,2], [3,4], [5,6], [7,8]] # lista de listas representando os dados de entrada
y = [[3], [7], [11], [15]] # lista de listas representando os valores de saida
```

```python
# conversão das listas x e y para tensores
X = torch.tensor(x).float()
Y = torch.tensor(y).float()
```

```python
# define o dispositivo de processamento usa gpu cuda
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(device)
```

```python
# transfere os tensores X e Y para o dispositivo
X = X.to(device)
Y = Y.to(device)
```

```python
# importação do modulo de redes neurais do pytorch
import torch.nn as nn
```

```python
# definição da classe para a rede neural customizada
class MyNeuralNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer1 = nn.Linear(2, 8)
        self.activation = nn.ReLU()
        self.layer2 = nn.Linear(8, 1)
    # definição do fluxo de dados
    def forward(self, x):
        x = self.layer1(x)
        x = self.activation(x)
        x = self.layer2(x)
        return x
```

```python
# instancia a rede neural
model = MyNeuralNet()
```
