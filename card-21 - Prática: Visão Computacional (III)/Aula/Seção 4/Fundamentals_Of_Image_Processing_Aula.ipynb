```python
# importando as bibliotecas
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

# garante que os gráficos e imagens
%matplotlib inline
```

```python
# lê a imagem baixada
img = cv.imread('ab6761610000e5eb31f6ab67e6025de876475814')
```

```python
# exibe a imagem usando
plt.imshow(img)
```

```python
rgb_image = cv.cvtColor(img, cv.COLOR_BGR2RGB) # converte o espaço de cores de BGR
plt.imshow(rgb_image) # exibe a imagem com as cores corrigidas
```

```python
gray_image = cv.cvtColor(img, cv.COLOR_BGR2GRAY) # converte a imagem colorida
plt.imshow(gray_image, cmap='gray') # exibe a imagem em cinza
```

```python
img_gray_small = cv.resize(gray_image, (50, 50)) # redimensiona a imagem para uma resolução fixa de 50x50
plt.imshow(img_gray_small, cmap='gray') # exibe a imagem redimensionada em tons de cinza
```

```python
# recorta a imagem mantendo as linhas de 10 a 340
plt.imshow(gray_image[10:340, :], cmap='gray')
```

```python
# recorta a imagem mantendo apenas ate a coluna 300
plt.imshow(gray_image[:, :300], cmap='gray')
```
