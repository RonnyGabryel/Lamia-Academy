```python
import math # importação da biblioteca matemática nativa

# implementação manual da função de ativação sigmoid
def sigmoid(x):
    return 1 / (1 + math.exp(-x))
```

```python
import torch # importação do pytorch para definir a classe

# implementação customizada da função de ativaçao
class ReLU(torch.nn.Module):
    def __init__(self):
        super().__init__()

    def forward(self, x):
        return torch.max(torch.zeros_like(x), x) # retorna o valor máximo entre zero e x
```

```python
import numpy as np # importação do numpy para calculos

# calculo do erro medio quadratico
def mse_loss(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)
```

```python
# cálculo da perda de entropia cruzada binária
def cross_entropy_loss(y_true, y_pred):
    epsilon = 1e-15
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
```
