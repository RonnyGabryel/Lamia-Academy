```python
# importando a biblioteca
import torch
import torch.nn as nn
import numpy as np
```

```python
# testando sigmoid com valores bem distantes
entrada = torch.tensor([-10.0, -5.0, 0.0, 5.0, 10.0])
ativacao_sigmoid = torch.sigmoid(entrada)
print(ativacao_sigmoid)
```

```python
# teste a função Tanh
tanh = nn.Tanh()
print(tanh(entrada))
```

```python
# calculo manual de MAE
def mae_loss(y_true, y_pred):
    return np.mean(np.abs(y_true - y_pred))
```

```python
y_real = np.array([1.0, 0.0, 1.0])
y_prev = np.array([0.9, 0.2, 0.8])
print("Erro médio:",  mae_loss(y_real, y_prev))
```
