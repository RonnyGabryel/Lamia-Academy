```python
# importação das bibliotecas de processamento de imagem
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# baixa uma imagem de exemplo da internet
!wget -q https://i.pinimg.com/736x/f9/82/67/f982676d44da212bf5c665e804b92bbf.jpg
```

```python
# le a imagem do disco
img = cv.imread('f982676d44da212bf5c665e804b92bbf.jpg')
print('Shape original:', img.shape)
```

```python
# converte BGR para RGB
rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(rgb)
plt.axis('off'); plt.show()
```

```python
# converte para escala de cinza
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
print('Shape em cinza:', gray.shape)
plt.imshow(gray, cmap='gray')
plt.title('Escala de cinza'); plt.axis('off'); plt.show()
```

```python
# redimensiona a imagem para 50x50
small = cv.resize(gray, (50, 50))
print('Shape redimensionado:', small.shape)
plt.imshow(small, cmap='gray')
plt.title('50x50'); plt.axis('off'); plt.show()
```

```python
# recorte por linhas
crop_rows = gray[10:340, :]
plt.imshow(crop_rows, cmap='gray')
plt.title(f'Recorte por linhas — shape: {crop_rows.shape}'); plt.axis('off'); plt.show()
```

```python
# recorte por colunas
crop_cols = gray[:, :300]
plt.imshow(crop_cols, cmap='gray')
plt.title(f'Recorte por colunas — shape: {crop_cols.shape}'); plt.axis('off'); plt.show()
```
