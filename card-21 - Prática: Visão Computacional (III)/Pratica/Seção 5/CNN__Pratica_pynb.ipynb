```python
# importando a biblioteca
import torch
import torch.nn as nn
from torch.optim import Adam
```

```python
# definindo o dispositivo
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
# criando dados de treino diferentes
x_train = torch.randn(2, 1, 6, 6).to(device)
y_train = torch.tensor([0, 1]).float().to(device)
```

```python
# arquitetura com Pooling e Padding
class CNNCustom(nn.Module):
    def __init__(self):
        super().__init__()
        self.camadas = nn.Sequential(
            nn.Conv2d(1, 4, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2),
            nn.Flatten(),
            nn.Linear(4 * 3 * 3, 1)
        )

    def forward(self, x):
        return self.camadas(x)

model = CNNCustom().to(device)
opt = Adam(model.parameters(), lr=0.005)
loss_fn = nn.BCEWithLogitsLoss()
```

```python
# loop de treino rápido para demonstrar
for epoch in range(100):
    opt.zero_grad()
    pred = model(x_train).squeeze()
    loss = loss_fn(pred, y_train)
    loss.backward()
    opt.step()

print(f"{loss.item():.4f}")
```
