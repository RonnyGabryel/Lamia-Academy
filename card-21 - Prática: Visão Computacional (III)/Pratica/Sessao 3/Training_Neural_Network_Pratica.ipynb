```python
# importação das bibliotecas principais
import torch
import torch.nn as nn
from torch.optim import SGD
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# definição do dataset
x = [[1,2], [3,4], [5,6], [7,8]]
y = [[3], [7], [11], [15]]
```

```python
# conversão das listas para tensores float
X = torch.tensor(x).float()
Y = torch.tensor(y).float()
```

```python
# configuraçao do dispositivo de hardware
device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
# transferencia dos tensores para o dispositivo
X = X.to(device)
Y = Y.to(device)
```

```python
# definiçao da arquitetura da rede neural
class MyNeuralNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer1 = nn.Linear(2, 8)
        self.activation = nn.ReLU()
        self.layer2 = nn.Linear(8, 1)

    # define o caminho do dado
    def forward(self, x):
        x = self.layer1(x)
        x = self.activation(x)
        x = self.layer2(x)
        return x

# instancia o modelo
model = MyNeuralNet().to(device)
```

```python
# define a função de perda
loss_fn = nn.MSELoss()
opt = SGD(model.parameters(), lr=0.001)
```

```python
# lista para armazenar o histórico
losses = []

# loop de treinamento por 50 range
for _ in range(50):
    opt.zero_grad() # zera o gradiente
    loss_value = loss_fn(model(X), Y) # calcula o erro
    loss_value.backward() # propaga o erro
    opt.step() # atualiza os pesos
    losses.append(loss_value.item())# salva a perda para plotar depois

# printa os erros
print(f"Erro inicial: {losses[0]:.2f}")
print(f"Erro final:   {losses[-1]:.2f}")
```
