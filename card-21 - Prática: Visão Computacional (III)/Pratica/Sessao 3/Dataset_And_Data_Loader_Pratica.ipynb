```python
# importando que serão usadas
from torch.utils.data import Dataset, DataLoader
import torch
import torch.nn as nn
```

```python
# dados de entrada e saída esperada
x = [[1,2], [3,4], [5,6], [7,8]]
y = [[3], [7], [11], [15]]
```

```python
# define o dispositivo
device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
# classe responsável para armazenar
class MyDataset(Dataset):
    def __init__(self, x, y):
        # converte as listas em tensores
        self.x = torch.tensor(x).float().to(device)
        self.y = torch.tensor(y).float().to(device)

    # retorna o número total de amostras
    def __len__(self):
        return len(self.x)

    # retorna uma amostra pelo índice
    def __getitem__(self, ix):
        return self.x[ix], self.y[ix]
```

```python
# instancia o dataset
ds = MyDataset(x, y)
print("Total de amostras:", len(ds))

# acessando uma amostra pelo índice
print("Amostra 0:", ds[0])
```

```python
# cria o dataloader
dl = DataLoader(ds, batch_size=2, shuffle=True)

# percorre os batches e exibe cada um
for xb, yb in dl:
    print("x:", xb.tolist(), " y:", yb.tolist())
```

```python
# definição da arquitetura
class MyNeuralNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer1 = nn.Linear(2, 8)
        self.activation = nn.ReLU()
        self.layer2 = nn.Linear(8, 1)

    # define o fluxo de dados
    def forward(self, x):
        x = self.layer1(x)
        x = self.activation(x)
        x = self.layer2(x)
        return x

# instancia o modelo
model = MyNeuralNet().to(device)
loss_fn = nn.MSELoss()
opt = torch.optim.SGD(model.parameters(), lr=0.001)
```

```python
# loop de treinamento
losses = []
for _ in range(50):
    for xb, yb in dl:# itera sobre os batches do dataloader
        opt.zero_grad() # zera o gradiente acumulado
        loss_value = loss_fn(model(xb), yb)  # calcula o erro do batch
        loss_value.backward() # propaga o erro
        opt.step()  # atualiza os pesos
        losses.append(loss_value.item())

print(f"Erro inicial: {losses[0]:.2f}")
print(f"Erro final:   {losses[-1]:.2f}")
```
