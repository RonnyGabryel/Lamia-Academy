```python
# importando as bibliotecas que seram necessarias
import torch
import torch.nn as nn
```

```python
# dados de entrada e saída esperada
x = [[1,2], [3,4], [5,6], [7,8]]
y = [[3], [7], [11], [15]]
```

```python
# conversão das listas para tensores float
X = torch.tensor(x).float()
Y = torch.tensor(y).float()
```

```python
# define o dispositivo
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(device)
```

```python
# move os tensores para o dispositivo
X = X.to(device)
Y = Y.to(device)
```

```python
# definição da arquitetura da rede neural com duas camadas
class MyNeuralNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer1 = nn.Linear(2, 8)  # entrada com 2 features para 8 neurônios
        self.activation = nn.ReLU()     # ativação entre as camadas
        self.layer2 = nn.Linear(8, 1)  # 8 neurônios para 1 saída

    # define o fluxo do dado pelas camadas
    def forward(self, x):
        x = self.layer1(x)
        x = self.activation(x)
        x = self.layer2(x)
        return x
```

```python
# instancia o modelo
model = MyNeuralNet().to(device)
# exibe os pesos iniciais da primeira camada
print("Pesos da layer:\n", model.layer1.weight)
# exibe o shape do tensor de entrada
print("\nShape de X:", X.shape)
```

```python
# passa os dados pela rede sem treinar
with torch.no_grad():
    pred = model(X)

print(pred)
```
