```python
# importando as bibliotecas
import math
import numpy as np
import torch
```

```python
# implementação manual da função sigmoid
def sigmoid(x):
    return 1 / (1 + math.exp(-x))
```

```python
# testando com valores negativos
for v in [-3, -1, 0, 1, 3]:
    print(f"sigmoid({v:2d}) = {sigmoid(v):.4f}")
```

```python
# implementação customizada da ReLU
class ReLU(torch.nn.Module):
    def __init__(self):
        super().__init__()

    # retorna zero para valores s
    def forward(self, x):
        return torch.max(torch.zeros_like(x), x)

# testando a relu com um tensor
relu = ReLU()
t = torch.tensor([-2.0, -1.0, 0.0, 1.0, 2.0])

# printa tudo
print("Entrada:", t.tolist())
print("Saída:  ", relu(t).tolist())
```

```python
# erro médio quadrático
def mse_loss(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

# testando com predições
y_true = np.array([1.0, 2.0, 3.0])
y_pred_bom  = np.array([1.1, 2.1, 3.1])
y_pred_ruim = np.array([2.0, 4.0, 0.0])

# printando as predição
print("MSE predição boa: ", mse_loss(y_true, y_pred_bom))
print("MSE predição ruim:", mse_loss(y_true, y_pred_ruim))
```

```python
# entropia cruzada binaria
def cross_entropy_loss(y_true, y_pred):
    epsilon = 1e-15
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))

# testando com predições
y_true = np.array([1, 0, 1, 1])
y_pred_bom  = np.array([0.9, 0.1, 0.8, 0.9])
y_pred_ruim = np.array([0.1, 0.9, 0.2, 0.1])

# printa a Cross Entropy
print("Cross Entropy boa: ", cross_entropy_loss(y_true, y_pred_bom))
print("Cross Entropy ruim:", cross_entropy_loss(y_true, y_pred_ruim))
```
