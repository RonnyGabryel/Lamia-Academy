```python
# importando a biblioteca
import torch
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
# baixa uma imagem do gato
!wget -q https://i.pinimg.com/736x/f9/82/67/f982676d44da212bf5c665e804b92bbf.jpg
```

```python
# carregando a foto do gato
img_path = 'f982676d44da212bf5c665e804b92bbf.jpg'
img = Image.open(img_path)
```

```python
# redimensionando e convertendo para numpy
img_res = img.resize((224, 224))
img_array = np.array(img_res)
```

```python
# separando os canais de cor RGB
canal_r = img_array[:, :, 0]
canal_g = img_array[:, :, 1]
canal_b = img_array[:, :, 2]
```

```python
# plotando os canais separadamente
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].imshow(canal_r, cmap='Reds')
axs[0].set_title("Canal Vermelho")
axs[1].imshow(canal_g, cmap='Greens')
axs[1].set_title("Canal Verde")
axs[2].imshow(canal_b, cmap='Blues')
axs[2].set_title("Canal Azul")
plt.show()
```

```python
# teste original: criando uma versão em tons de cinza
img_gray = np.mean(img_array, axis=2)
plt.imshow(img_gray, cmap='gray')
plt.title("Gato em Tons de Cinza (Média Manual)")
plt.show()
```
