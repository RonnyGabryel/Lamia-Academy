```python
# importando a biblioteca
import torch
import torch.nn as nn
from torch.optim import RMSprop # coloquei um diferente otimizador
```

```python
class RedeNeuralCustom(nn.Module):
    def __init__(self):
        super().__init__()
        self.sequencial = nn.Sequential(
            nn.Linear(28 * 28, 512),
            nn.BatchNorm1d(512),
            nn.ReLU(),
            nn.Linear(512, 128),
            nn.ReLU(),
            nn.Linear(128, 10)
        )

    def forward(self, x):
        # achatando a imagem de 28x28 para um vetor de 784
        x = x.view(-1, 28 * 28)
        return self.sequencial(x)
```

```python
# instanciando o modelo
model = RedeNeuralCustom()
```

```python
# usando RMSprop com uma taxa de aprendizado
optimizer = RMSprop(model.parameters(), lr=0.001)
loss_fn = nn.CrossEntropyLoss()

print(model)
```

```python
# teste de shape para garantir
input_teste = torch.randn(1, 28, 28)

# Colocar o modelo em modo de avaliação
model.eval()
output = model(input_teste)
print("shape da saida:", output.shape)
```
