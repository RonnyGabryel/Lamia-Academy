```python
# biblioteca torch
import torch
```

```python
# criando um tensor 5x2 com valores decimais
x = torch.tensor([[1.5, 2.5], [3.5, 4.5], [5.5, 6.5], [7.5, 8.5], [9.5, 10.5]])
print(x)
print("Shape original:", x.shape)
```

```python
# multiplicando por um valor decimal e somando um valor inteiro
print(x * 0.5)
print(x.add(100))
```

```python
# mudando o formato com view para 2 linhas e 5 colunas
print("\nview(2, 5):\n", x.view(2, 5))
```

```python
# criando tensor com dimensões extras para testar
a = torch.ones(1, 5, 1, 2)
print(a.shape)
print(a.squeeze().shape)
print(a.squeeze(0).shape)
```

```python
# criando matriz y que combine com o shape de x
y = torch.tensor([[2, 1, 0],
                  [0, 1, 2]], dtype=torch.float)
resultado = torch.matmul(x, y)
print(resultado)
print(resultado.shape)
```

```python
# testando concatenação em eixos diferentes
b = torch.randn(3, 2)
cat_dim0 = torch.cat([b, b], dim=0)
cat_dim1 = torch.cat([b, b], dim=1)
print(b.shape)
print(cat_dim0.shape)
print(cat_dim1.shape)
```

```python
# invertendo a ordem dos eixos
c = torch.randn(2, 10, 5)
print(c.shape)
print(c.permute(2, 0, 1).shape)
```
