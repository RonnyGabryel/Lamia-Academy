## Criando e analisando um sinal

Nesta prática, vamos aplicar o que foi mostrado na aula para
1. Gerar um conjunto de dados com `numpy`
2. Criar uma função senoidal modificada
3. Plotar o gráfico usando `matplotlib`

```python
print("Iniciando prática do Google Colab")
```

```python
### Importando bibliotecas necessárias
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Gerando dados
t = np.arange(0.0, 4.0, 0.01)
s = 2 * np.sin(2 * np.pi * 1.5 * t)

# Criando o gráfico
fig, ax = plt.subplots()
ax.plot(t, s, color='purple', linewidth=2)
ax.grid()
plt.show()
```
