```python
# Importando bibliotecas necessárias
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Gerando 200 valores entre 0 e 4π
x = np.linspace(0, 4 * np.pi, 200)

# Função seno com ruído aleatório
y = np.sin(x) + np.random.normal(0, 0.2, x.size)

# Criando o gráfico
fig, ax = plt.subplots(figsize=(7, 4))
ax.plot(x, y, color='pink', linewidth=2, label='seno com ruído')

# Configurando o gráfico
ax.set_title("Visualizando dados simulados")
ax.set_xlabel("Ângulo (rad)")
ax.set_ylabel("Amplitude")
ax.legend()
ax.grid(True)

plt.show()
```
