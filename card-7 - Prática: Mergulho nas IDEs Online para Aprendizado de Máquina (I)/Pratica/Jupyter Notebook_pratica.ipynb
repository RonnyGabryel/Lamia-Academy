```python
# Importando as bibliotecas
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
# Criando 100 valores igualmente espaçados entre 0 e 4
x = np.linspace(0, 4 * np.pi, 100)

# Calculando cosseno de x com modificação de amplitude
y = 1.5 * np.cos(x) + 0.5

# Criando figura
fig, ax = plt.subplots()

# Plotando a linha
ax.plot(x, y, color='blue', linewidth=2)

# Definindo rótulos
ax.set(xlabel='ângulo (rad)', ylabel='amplitude',
       title='Visualização de dados é fundamental')
ax.grid()
plt.show()
```
