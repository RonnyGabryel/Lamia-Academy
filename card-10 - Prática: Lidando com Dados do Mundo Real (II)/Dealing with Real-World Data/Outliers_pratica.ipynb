```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

# Activity

Instead of a single outlier, add several randomly-generated outliers to the data. Experiment with different values of the multiple of the standard deviation to identify outliers, and see what effect it has on the final results.

```python
np.random.seed(100)

# Dados normais
incomes = np.random.normal(27000, 15000, 10000)
outliers = np.random.normal(70000, 30000, 1000)

incomes = np.append(incomes, outliers)

# Visualiza dados com outliers
plt.hist(incomes, 50)
plt.show()

incomes.mean()
```

```python
# Função que remove outliers baseado em múltiplos do desvio padrão
def reject_outliers(data, m=2):
    u = np.median(data)
    s = np.std(data)
    filtered = [e for e in data if (u - m * s < e < u + m * s)]
    return filtered

# Filtra outliers com m=2
filtered = reject_outliers(incomes)

plt.hist(filtered, 50)
plt.show()

np.mean(filtered)
```

```python
# Testa diferentes valores de m
for i, m in enumerate([0.5, 1, 1.5, 2, 2.5, 3]):
    filtered = reject_outliers(incomes, m)
    plt.subplot(3, 2, i + 1)
    plt.subplots_adjust(hspace=0.8)
    plt.hist(filtered, 50)
    plt.title('m = {}'.format(m))
    print('m = {}: {}'.format(m, np.mean(filtered)))
plt.show()
```
