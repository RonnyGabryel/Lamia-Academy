# Sigmoid
O que faz: "Espreme" qualquer valor para um intervalo entre 0 e 1.

```python
 def sigmoid_function(soma):
     return 1 / (1 + np.exp(-soma))
```

```python
sigmoid_function(-10)
```

```python
sigmoid_function(10)
```

```python
sigmoid_function(1)
```

# Hyperbolic
O que faz: Transforma os valores para o intervalo entre -1 e 1

```python
def tah_function(soma):
    return (np.exp(soma) - np.exp(soma)) / (np.exp(soma) + np.exp(-soma))
```

```python
tah_function(-10)
```

```python
tah_function(10)
```

```python
tah_function(1)
```

# ReLU
O que faz: Se o número for negativo, vira 0 se for positivo ele permanece igual

Objetivo: Eliminar valores negativos e manter a rede eficiente

```python
def relu_function(soma):
    if soma >= 0:
     return soma
    return 0
```

```python
relu_function(-20)
```

```python
relu_function(0)
```

```python
relu_function(10)
```

```python
relu_function(100)
```

# Softmax
O que faz: Transforma um grupo de números em probabilidades que somam 100\% (1.0.

Objetivo: Indicar qual classe é a mais provável "tem 97% de chance de ser um gato"

```python
def softmax_function(x):
    ex = np.exp(x)
    return ex / ex.sum()
```

```python
softmax_function([7.0, 2.0, 1,3])
```

```python
softmax_function([2.0, 2.0, 5,3])
```
