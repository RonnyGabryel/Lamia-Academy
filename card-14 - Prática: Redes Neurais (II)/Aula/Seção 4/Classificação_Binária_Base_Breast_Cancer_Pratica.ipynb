```python
# importando as bibliotecas
import pandas as pd
import tensorflow as tf
from scikeras.wrappers import KerasClassifier
from tensorflow.keras import backend as K
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, InputLayer
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
```

```python
def criar_rede(optimizer, kernel_initializer, neurons):
    K.clear_session()  # limpa
    rede_neural = Sequential([
        tf.keras.layers.InputLayer(shape=(30,)),# camada de entrada 30 features        
        tf.keras.layers.Dense(units=neurons, activation='relu', kernel_initializer=kernel_initializer),#Camada Oculta
        tf.keras.layers.Dropout(rate=0.2), # Regularização
        tf.keras.layers.Dense(units=neurons, activation='relu', kernel_initializer=kernel_initializer),#Camada Oculta mesma estrutura para profundidade
        tf.keras.layers.Dropout(rate=0.2),        
        tf.keras.layers.Dense(units=1, activation='sigmoid')#Camada de saída ativação Sigmoid para classificação binária 0 ou 1
    ])
    # Configuração do treino
    rede_neural.compile(
        optimizer=optimizer,
        loss='binary_crossentropy',
        metrics=['binary_accuracy']
    )
    return rede_neural
```

```python
# Configuração do Wrapper
rede_neural = KerasClassifier(model=criar_rede, verbose=0)
```
