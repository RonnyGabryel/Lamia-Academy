```python
from scikeras.wrappers import KerasClassifier #importando o KerasClassifier
from sklearn.model_selection import cross_val_score #importando o cross_val_score
from tensorflow.keras.models import Sequential #importando o Sequential
from tensorflow.keras import backend as K #importando o backend
```

Artigo dropout: https://www.cs.toronto.edu/~rsalakhu/papers/srivastava14a.pdf

```python
#importando o arquivo de entradas
x = pd.read_csv('/home/Ront/Documentos/Card-14/Seção 4/entradas_breast.csv') 
y = pd.read_csv('/home/Ront/Documentos/Card-14/Seção 4/saidas_breast.csv') 
```

```python
def criar_rede(): #criando a função para criar a rede neural
         K.clear_session() #limpando a sessão do Keras
         rede_neural = Sequential([ #criando o modelo sequencial
             tf.keras.layers.InputLayer(shape=(30,)), #camada de entrada com 30 features
             tf.keras.layers.Dense(units=16, activation='relu', kernel_initializer='random_uniform'), #primeira camada oculta com 16 neurônios
             tf.keras.layers.Dense(units=16, activation='relu', kernel_initializer='random_uniform'), #segunda camada oculta com 16 neurônios
             tf.keras.layers.Dense(units=1, activation='sigmoid') #camada de saída com 1 neurônio para classificação binária
         ])
         otimizador = tf.keras.optimizers.Adam(learning_rate=0.001, clipvalue=0.5) #definindo o otimizador Adam
         rede_neural.compile(optimizer=otimizador, loss='binary_crossentropy', metrics=['binary_accuracy']) #compilando a rede
         return rede_neural #retornando a rede criada
```

```python
rede_neural = KerasClassifier(model=criar_rede, epochs=100, batch_size=10)
```

```python
resultados = cross_val_score(estimator=rede_neural, X=x, y=y, cv=10, scoring='accuracy')
```

# 27. Overfitting e dropout - implementação

```python
def criar_rede(): #criando a função para criar a rede neural
    K.clear_session() #limpando a sessão do Keras
    rede_neural = Sequential([ #criando o modelo sequencial
        tf.keras.layers.InputLayer(shape=(30,)), #camada de entrada com 30 features
        tf.keras.layers.Dense(units=16, activation='relu', kernel_initializer='random_uniform'), #primeira camada oculta com 16 neurônios
        tf.keras.layers.Dropout(rate=0.2), #dropout de 20% para evitar overfitting
        tf.keras.layers.Dense(units=16, activation='relu', kernel_initializer='random_uniform'), #segunda camada oculta com 16 neurônios
        tf.keras.layers.Dropout(rate=0.2), #dropout de 20% para evitar overfitting
        tf.keras.layers.Dense(units=1, activation='sigmoid') #camada de saída com 1 neurônio para classificação binária
    ])
    otimizador = tf.keras.optimizers.Adam(learning_rate=0.001, clipvalue=0.5) #definindo o otimizador Adam
    rede_neural.compile(optimizer=otimizador, loss='binary_crossentropy', metrics=['binary_accuracy']) #compilando a rede
    return rede_neural #retornando a rede criada
```

```python
rede_neural = KerasClassifier(model=criar_rede, epochs=100, batch_size=10)
```

```python
resultados = cross_val_score(estimator=rede_neural, X=x, y=y, cv=10, scoring='accuracy')
```
