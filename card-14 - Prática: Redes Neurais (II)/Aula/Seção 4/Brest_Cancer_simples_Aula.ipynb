```python
#criando uma instância da classe Sequential
rede_neural = Sequential([

  tf.keras.layers.InputLayer(shape = (30,)), # Define que a rede receberá 30 atributos colunas de entrada
  tf.keras.layers.Dense(units = 16, activation = 'relu', kernel_initializer = 'random_uniform'),  # Camada oculta com 16 neurônio função ReLU e pesos iniciais aleatórios
  tf.keras.layers.Dense(units = 16, activation = 'relu', kernel_initializer = 'random_uniform'),  # Camada oculta com 16 neurônio função ReLU e pesos iniciais aleatórios
  tf.keras.layers.Dense(units = 1, activation = 'sigmoid')# Camada de saída com 1 neurônio e função Sigmoid ideal para classificação binária
])
```
