```python
#Importandao as bibliotecas que ira ser usadas
import pandas as pd
import tensorflow as tf
import sklearn
import scikeras
```

```python
from scikeras.wrappers import KerasClassifier
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras import utils as np_utils
from tensorflow.keras import backend as k
from tensorflow.keras.models import Sequential
from sklearn.model_selection import cross_val_score
```

```python
# Pegando os dados
base = pd.read_csv('/home/Ront/Documentos/Card-14/Seção 5/iris.csv') 
X = base.iloc[:, 0:4].values
y = base.iloc[:, 4].values
```

```python
labelencoder = LabelEncoder()
y = labelencoder.fit_transform(y) # Conversão para valores numéricos
y = np_utils.to_categorical(y) # Conversão para um vetor com a posição
```

```python
# Criando a rede
def criar_rede():
    k.clear_session()
    rede_neural = Sequential([
        tf.keras.layers.InputLayer(shape=(4,)),
        tf.keras.layers.Dense(units=4, activation='relu'),
        tf.keras.layers.Dense(units=4, activation='relu'),
        tf.keras.layers.Dense(units=3, activation='softmax')])
    rede_neural.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['categorical_accuracy'])
    return rede_neural
```

```python
rede_neural = KerasClassifier(model = criar_rede, epochs = 250, batch_size = 10)
```

```python
resultados = cross_val_score(estimator=rede_neural, X = X, y = y, cv = 10, scoring = 'accuracy')
```
