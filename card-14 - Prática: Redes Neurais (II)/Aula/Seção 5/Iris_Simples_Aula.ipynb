```python
# Importando as bibliotecas
import pandas as pd
import numpy as np
import tensorflow as tf
import sklearn
```

```python
pd.__version__, np.__version__, tf.__version__, sklearn.__version__
```

```python
# Importa o tipo de modelo mais comum uma pilha linear de camadas
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense #Importa a camada "Densa", que é a camada neural básica onde 
from tensorflow.keras import utils as np_utils# Importa utilitários do Keras como o to_categorical para converter 
from sklearn.preprocessing import LabelEncoder# Ferramenta do Scikit-Learn para converter rótulos em formato de texto 
from sklearn.model_selection import train_test_split# Função crucial para dividir seu conjunto de dados: uma parte para o 
from sklearn.metrics import confusion_matrix
```

```python
base = pd.read_csv('/home/Ront/Documentos/Card-14/Seção 5/iris.csv') 
base
```

```python
X = base.iloc[:, 0:4].values # Atributos previsores 150 linhas e 4 colunas
X
```

```python
y = base.iloc[:, 4].values # Atributos classe 150 linhas e 1 coluna
y
```

```python
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y) #transforma as labels em números
y
```

```python
y.shape
```

```python
y = np_utils.to_categorical(y) # Transforma os valores de y em categorias 
y
```

```python
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(X, y, test_size=0.25)
```

```python
X_treinamento.shape, X_teste.shape
```

```python
y_treinamento.shape, y_teste.shape
```

```python
rede_neural = Sequential([
    tf.keras.layers.InputLayer(shape = (4,)), # Camada de entrada com 4 neurônios
    tf.keras.layers.Dense(units = 4, activation = 'relu'),
    tf.keras.layers.Dense(units = 4, activation = 'relu'),
    tf.keras.layers.Dense(units = 3, activation = 'softmax')
    ])
```

```python
rede_neural.summary()
```

```python
# Função de perda e métricas
rede_neural.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['categorical_accuracy'])
```

```python
rede_neural.fit(X_treinamento, y_treinamento, batch_size = 10, epochs = 1000)
```

```python
rede_neural.evaluate(X_teste, y_teste) # Avalia o modelo com os dados
```

```python
previsoes = rede_neural.predict(X_teste) # Faz as previsões com os dados de teste
print(previsoes)
```

```python
previsoes = previsoes > 0.5
print(previsoes)
```

```python
y_teste2 = [np.argmax(t) for t in y_teste] # Retorna o índice do valor True em cada linha
print(y_teste2)
```

```python
previsoes2 = [np.argmax(t) for t in previsoes]
print(previsoes2)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_teste2, previsoes2)
```

```python
confusion_matrix(y_teste2, previsoes2)
```
