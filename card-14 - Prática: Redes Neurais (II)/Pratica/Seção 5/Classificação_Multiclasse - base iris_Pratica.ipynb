```python
#Importando oq vai ser usado
import pandas as pd
import tensorflow as tf
from scikeras.wrappers import KerasClassifier
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras import utils as np_utils
from tensorflow.keras import backend as K
from tensorflow.keras.models import Sequential
from sklearn.model_selection import cross_val_score
```

```python
# Carregar dados
base = pd.read_csv('/home/Ront/Documentos/Card-14/Pratica/Seção 5/iris.csv') 
X = base.iloc[:, 0:4].values
y = base.iloc[:, 4].values
```

```python
# Codificar as classes
labelencoder = LabelEncoder()
y = labelencoder.fit_transform(y)
y = np_utils.to_categorical(y)
```

```python
#CRiando a rede
def criar_rede():
    K.clear_session()
    
    modelo = Sequential([
        tf.keras.layers.InputLayer(shape=(4,)),
        tf.keras.layers.Dense(units=4, activation='relu'),
        tf.keras.layers.Dense(units=3, activation='softmax')
    ])
    
    modelo.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['categorical_accuracy'])
    return modelo
```

```python
# Criar modelo com treinamento simplificado
modelo_keras = KerasClassifier(model=criar_rede, epochs=100, batch_size=5)
```

```python
# Validação cruzada com 5 folds
resultados = cross_val_score(estimator=modelo_keras, X=X, y=y, cv=10, scoring='accuracy')
```
