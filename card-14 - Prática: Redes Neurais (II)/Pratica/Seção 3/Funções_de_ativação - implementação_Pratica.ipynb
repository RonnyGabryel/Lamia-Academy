# Step Function

```python
# Define a função que recebe a 'soma' dos sinais do neurônio
def step_function(soma):
    if soma >= 1:    # Se a soma atingir o limite 1 o neurônio ativa
        return 1     # Retorna 1 
    return 0         # Caso contrári, retorna 0 
```

```python
# Testes:
print(step_function(-10)) # Resultado: 0 Abaixo do limite
print(step_function(10))  # Resultado: 1 Acima do limite
print(step_function(1))   # Resultado: 1 Exatamente no limite
```

# Sigmoid

```python
#Define a função matemática da Sigmóide
def sigmoid_function(soma):
    #A fórmula utiliza o número de Euler np.exp
    return 1 / (1 + np.exp(-soma))
```

```python
# Testes:
# Transforma um número muito baixo -10 em algo quase 0
print(sigmoid_function(-10)) 

# Transforma um número alto 10 em algo quase 1 0.999...
print(sigmoid_function(10))
```

# Hyperbolic

```python
# Define a função tangente hiperbólica
def tah_function(soma):
    # Faz a relação entre exponenciais positivos e negativos
    return (np.exp(soma) - np.exp(soma)) / (np.exp(soma) + np.exp(-soma))
```

```python
# Testes:
print(tah_function(-10)) # retorna 0.0
print(tah_function(10))  #retorna 0.0
```

# ReLU

```python
def relu_function(soma):
    if soma >= 0:
        return soma # Se for positivo retorna o próprio valor
    return 0        # Se for negativo bloqueia o sinal retornando zero
```

```python
# Teste
print(relu_function(-20))  # Resultado: 0
print(relu_function(100))  # Resultado: 100
```

# Softmax

```python
def softmax_function(x):
    ex = np.exp(x)       # Eleva os valores à potência exponencial para destacar os maiores
    return ex / ex.sum() # Divide cada valor pela soma total para gerar o percentual

#teste 
    #Entrada [7.0, 2.0, 1.3] resulta em probabilidades como 0.97 (97%)
```
